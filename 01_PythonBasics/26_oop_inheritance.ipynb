# 26. Object-Oriented Programming: Inheritance  
Welcome! In this notebook, we’ll learn about **inheritance** – a powerful way to create new classes based on existing ones.  
Inheritance helps us model real‑world relationships like “a Science student **is a** student” and reuse code so we don’t have to write everything from scratch.

We’ll use the example of students in **Plus Two (12th class)** with different streams: Science, Commerce, and Humanities. Each stream has common attributes (name, roll number, school) but also some stream‑specific subjects and calculations.

---

## 1. What is Inheritance?  

- **Inheritance** lets you create a new class (child) that automatically gets all the attributes and methods of an existing class (parent).  
- The **parent class** (also called base class) contains the common code.  
- The **child class** (subclass) can add new features or change (override) existing ones.  
- This promotes **code reuse** – write once, use in many places.

> 👨‍👦 **Analogy**: A child inherits traits from parents but also has their own unique characteristics


---

## 2. Creating a Parent Class – `Student`  

Let’s start with a general `Student` class that holds common data for any Plus Two student.  
We’ll keep it simple: name, roll number, school, and a list of marks for common subjects (say, English and a second language).

```python
class Student:
    def __init__(self, name, roll_no, school, marks):
        """marks is a list of marks for common subjects"""
        self.name = name
        self.roll_no = roll_no
        self.school = school
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def percentage(self):
        total_possible = len(self.marks) * 100
        return (self.total_marks() / total_possible) * 100

    def display_info(self):
        print(f"Name   : {self.name}")
        print(f"Roll No: {self.roll_no}")
        print(f"School : {self.school}")
        print(f"Marks  : {self.marks}")
        print(f"Total  : {self.total_marks()}")
        print(f"Percentage: {self.percentage():.2f}%")
```

Now create a generic student (though in reality, every student belongs to a stream – but this class can still be used).

```python
# Example: a generic student with two common subjects
generic = Student("Rahul", 101, "GHSS Trivandrum", [85, 90])  # English, Second Language
generic.display_info()
```

---

## 3. Creating a Child Class – `ScienceStudent`  

Now let’s create a subclass for **Science** students. A Science student has all the traits of a general student, plus extra subjects like Physics, Chemistry, and Mathematics.  
We’ll also add a method to calculate the science total.

```python
class ScienceStudent(Student):          # Inherit from Student
    def __init__(self, name, roll_no, school, marks, science_marks):
        """
        marks: list of marks for common subjects (e.g., English, Second Lang)
        science_marks: list of marks for Physics, Chemistry, Maths (in that order)
        """
        # Call the parent class __init__ to set common attributes
        super().__init__(name, roll_no, school, marks)
        self.science_marks = science_marks   # stream-specific attribute

    def total_marks(self):
        # Override parent method to include science marks
        return super().total_marks() + sum(self.science_marks)

    def display_info(self):
        # Override to show science subjects too
        super().display_info()
        print("Science Subjects:")
        subjects = ["Physics", "Chemistry", "Mathematics"]
        for sub, mark in zip(subjects, self.science_marks):
            print(f"  {sub}: {mark}")
```

```python
# Common subjects: English (85), Malayalam (90)
# Science marks: Physics (92), Chemistry (88), Maths (95)
science_stu = ScienceStudent("Lakshmi", 102, "GHSS Trivandrum", [85, 90], [92, 88, 95])
science_stu.display_info()
```


Notice:  
- We reused the parent’s `__init__` via `super()` to set common attributes.  
- We added a new attribute `science_marks`.  
- We **overrode** `total_marks()` and `display_info()` to include science marks.  
- The inherited `percentage()` method still works correctly because it uses the overridden `total_marks()`.

## 4. Another Child Class – `CommerceStudent`  

Now let’s create a subclass for Commerce students, with subjects like Accountancy, Business Studies, and Economics.


```python
class CommerceStudent(Student):
    def __init__(self, name, roll_no, school, marks, commerce_marks):
        super().__init__(name, roll_no, school, marks)
        self.commerce_marks = commerce_marks   # list: [Accountancy, Business Studies, Economics]

    def total_marks(self):
        return super().total_marks() + sum(self.commerce_marks)

    def display_info(self):
        super().display_info()
        print("Commerce Subjects:")
        subjects = ["Accountancy", "Business Studies", "Economics"]
        for sub, mark in zip(subjects, self.commerce_marks):
            print(f"  {sub}: {mark}")
```

```python
commerce_stu = CommerceStudent("Arjun", 103, "GHSS Trivandrum", [80, 88], [91, 85, 78])
commerce_stu.display_info()
science_stu2 = ScienceStudent("Nikhil", 104, "GHSS Trivandrum", [90, 89], [92, 88, 95])
```

## 5. Using `isinstance()` and `issubclass()`  

Python provides built‑in functions to check relationships between objects and classes.


```python
# Check if science_stu2 is an instance of ScienceStudent
print(isinstance(science_stu2, ScienceStudent))   # True
print(isinstance(science_stu2, Student))          # True (because ScienceStudent inherits from Student)
print(isinstance(science_stu2, CommerceStudent))  # False

# Check if ScienceStudent is a subclass of Student
print(issubclass(ScienceStudent, Student))        # True
print(issubclass(CommerceStudent, Student))       # True
print(issubclass(Student, ScienceStudent))        # False
```

## 8. Benefits of Inheritance – Code Reuse  

- The common code (name, roll_no, school, basic methods) is written only once in the parent `Student` class.  
- Each child class adds only what is specific to that stream.  
- If we need to change how we store names or calculate percentages, we change it in one place (the parent) and all children automatically get the update.  
- This makes programs **easier to maintain and extend**.

## Summary  

- **Inheritance** allows a child class to acquire attributes and methods from a parent class.  
- Use `class Child(Parent):` to create a subclass.  
- Call the parent’s `__init__` using `super().__init__(...)` to initialise inherited attributes.  
- You can **override** methods in the child to change behaviour.  
- You can add new methods and attributes specific to the child.  
- Inheritance models **“is‑a”** relationships (e.g., ScienceStudent **is a** Student).  
- It promotes **code reuse** and makes your code more organised.
