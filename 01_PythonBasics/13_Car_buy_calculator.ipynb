## 📊 The Strategy: The 20/4/10 Rule 

Before coding, let's understand the **financial strategy** behind the calculator:

### The 28/36 Rule (Simplified to 20% for EMI)
We uses a **conservative 20% of monthly salary for EMI**. This is stricter than the standard 28% guideline, ensuring you have room for other expenses.

### Our Calculation Logic:
1. **Max EMI** = Monthly Salary × (EMI Percentage / 100)
2. **Loan Amount** = Calculated using the loan formula:  
   `Loan = EMI × [(1 + r)^n - 1] / [r × (1 + r)^n]`  
   (where r = monthly interest rate, n = months)
3. **Vehicle Price** = Loan Amount + Down Payment
4. **Down Payment** = Vehicle Price - Loan Amount (we solve for this)

### Key Assumptions:
- 20% of salary is the maximum recommended for EMI
- 7% annual interest rate (typical for car loans in India)
- 5-year loan tenure (60 months)
- You should have at least 20% down payment saved

Now let's build this step by step!

```python
# Get inputs from user
monthly_salary = 50000
emi_percent = 15
interest_rate = 7
loan_years = 5

# Display what was entered
print("\n--- Your Inputs ---")
print(f"Monthly Salary: ₹{monthly_salary:,.2f}")
print(f"EMI Percentage: {emi_percent}%")
print(f"Interest Rate: {interest_rate}% per year")
print(f"Loan Tenure: {loan_years} years")
```

```python
# Calculate max EMI
max_emi = monthly_salary * (emi_percent / 100)
print(f"Maximum monthly EMI: ₹{max_emi:,.2f}")
```

## Calculate Maximum Loan Amount

This requires the **loan EMI formula**. Don't worry – we'll break it down:

**Formula:**  
`Loan = EMI × [(1 + r)^n - 1] / [r × (1 + r)^n]`

Where:
- `r` = monthly interest rate (annual rate ÷ 12 ÷ 100)
- `n` = total months (years × 12)

Let's implement this step by step.

```python
# Convert annual interest rate to monthly decimal
monthly_rate = interest_rate / (12 * 100)
total_months = loan_years * 12

# Calculate loan amount using the formula
# (1 + r)^n
factor = (1 + monthly_rate) ** total_months

# Loan = EMI × (factor - 1) / (monthly_rate × factor)
max_loan = max_emi * (factor - 1) / (monthly_rate * factor)

print(f"Maximum loan amount: ₹{max_loan:,.2f}")
```

```python
# Minimum down payment percentage (typical Indian bank requirement)
min_down_percent = 20

# Calculate minimum down payment based on loan amount
# If loan is 80% of price, then price = loan / 0.8
# And down payment = price - loan = loan × (0.2/0.8) = loan × 0.25
min_down_payment = max_loan * (min_down_percent / (100 - min_down_percent))

# Maximum vehicle price
max_vehicle_price = max_loan + min_down_payment

print(f"Minimum down payment (20% of vehicle price): ₹{min_down_payment:,.2f}")
print(f"Maximum vehicle price you can afford: ₹{max_vehicle_price:,.2f}")
```

```python
def format_number(n):
    s = str(int(n))
    if len(s) <= 3:
        return s
    
    # Take the last 3 digits
    last_three = s[-3:]
    # Take everything before the last 3 digits
    remaining = s[:-3]
    
    # Reverse the remaining string to group by 2s from right to left
    remaining_reversed = remaining[::-1]
    groups = []
    for i in range(0, len(remaining_reversed), 2):
        groups.append(remaining_reversed[i:i+2])
    
    # Join the 2-digit groups with commas, reverse back, and add the last 3
    formatted_remaining = ",".join(groups)[::-1]
    return f"{formatted_remaining},{last_three}"
parse_int = lambda val: int(str(val).replace(",", ""))
```

```python
def car_affordability(monthly_salary, emi_percent, interest_rate, loan_years):
    """
    Calculate max vehicle price, loan amount, down payment, and EMI
    based on salary and loan parameters.
    """
    # Step 1: Max EMI
    max_emi = monthly_salary * (emi_percent / 100)
    
    # Step 2: Monthly interest rate and total months
    monthly_rate = interest_rate / (12 * 100)
    total_months = loan_years * 12
    
    # Step 3: Loan amount calculation
    factor = (1 + monthly_rate) ** total_months
    max_loan = max_emi * (factor - 1) / (monthly_rate * factor)
    
    # Step 4: Minimum down payment (20% of vehicle price)
    min_down_payment = max_loan * 0.25  # 20% down means loan is 80% of price → down = loan × 0.25
    
    # Step 5: Max vehicle price
    max_price = max_loan + min_down_payment
    
    return {
        "max_emi": format_number(max_emi),
        "max_loan": format_number(max_loan),
        "min_down_payment": format_number(min_down_payment),
        "max_price": format_number(max_price)
    }

# Test with our example
result = car_affordability(100000, 20, 7, 5)
print("--- Calculation Results ---")
print(f"Maximum Monthly EMI: ₹{result['max_emi']}")
print(f"Maximum Loan Amount: ₹{result['max_loan']}")
print(f"Minimum Down Payment: ₹{result['min_down_payment']}")
print(f"Maximum Vehicle Price: ₹{result['max_price']}")
```
