# Loan Prepayment Impact Analyzer

In this notebook we will build a **Loan Prepayment Impact Analyzer** step by step.

The program will simulate a **30-year loan** and calculate how early the loan finishes if we:

1. Pay **extra money every year**
2. Pay **extra money in the last two months of every year**

This uses the **Reduced Balance Method**, meaning interest is calculated only on the remaining loan balance.

We will build this using only:

* Variables
* Arithmetic operations
* Lists & dictionaries
* `for` loops
* `if` conditions
* Basic built-in functions

No external libraries are used.

---

# Why This Matters

Understanding loan prepayment helps answer questions like:

* If I pay **extra every year**, how many years will I save?
* How much faster will the loan finish?

This is very useful for real life when dealing with:

* Home loans
* Car loans
* Personal loans

---

First we define the basic loan details.

````python
# Loan details
k = 1000
lakh = 100*k

loan_amount = 30*lakh
annual_interest_rate = 0.092
loan_years = 30

months = loan_years * 12

# Monthly interest rate
monthly_rate = annual_interest_rate / 12

print("Loan Amount:", loan_amount)
print("Loan Years:", loan_years)
print("Total Months:", months)
````

# Calculate Monthly EMI

To simulate a loan we first need the **EMI (Equated Monthly Installment)**.

Formula:

```
EMI = P * r * (1+r)^n / ((1+r)^n - 1)
```

Where

* **P** = loan amount
* **r** = monthly interest rate
* **n** = total months


````python
P = loan_amount
r = monthly_rate
n = months

emi = P * r * (1+r)**n / ((1+r)**n - 1)
print(emi)
````

# Track Remaining Loan Balance

Now we simulate **how the loan reduces every month**.

Each EMI contains:

* Interest part
* Principal repayment

Formula:

```
interest = balance * rate
principal_paid = emi - interest
balance = balance - principal_paid
```

````python
balance = loan_amount
print(f"{'Year':<4} | {'Interest Paid':>15} | {'Principal Paid':>20} | {'Balance':>10}")
print("-" * 65)
for month in range(0, months+1):

    interest = balance * monthly_rate
    principal_paid = emi - interest
    balance = balance - principal_paid
    
    if month%12 == 0 :
        print(f"{month/12:<4.1f} | {interest:>15,.2f} | {principal_paid:>20,.2f} | {balance:>10,.2f}")
````

# Add Yearly Extra Payment

Many people make **extra payments once per year**.


````python
print(f"{'Year':<4} | {'Interest Paid':>15} | {'Principal Paid':>20} | {'Balance':>10}")
print("-" * 65)

balance = loan_amount
extra_yearly_payment = 10 * k

for month in range(0, months+1):

    interest = balance * monthly_rate
    principal_paid = emi - interest
    balance = balance - principal_paid

    # every 12 months apply extra payment
    if month % 12 == 0:
        balance = balance - extra_yearly_payment
        print(f"{month/12:<4.1f} | {interest:>15,.2f} | {principal_paid:>20,.2f} | {balance:>10,.2f}")
    if balance< 0 : 
        break
print(f"{month/12:<4.1f} | {interest:>15,.2f} | {principal_paid:>20,.2f} | {balance:>10,.2f}")
print("-" * 65)
print(f"Loan will be finished in {month//12} year {month%12} months")
print("-" * 65)
````

````python
# loan amount pre payment
print(f"{'Year':<4} | {'Interest Paid':>15} | {'Principal Paid':>20} | {'Balance':>10}")
print("-" * 65)

balance = loan_amount
extra_yearly_payment = 10* k*0
extra_month_payment = 2*k
for year in range(0, months//12):
    for month in range(0,12):
        if balance< 0 : 
            break
        interest = balance * monthly_rate
        principal_paid = emi - interest
        balance = balance - principal_paid
        balance = balance - extra_month_payment
    if balance< 0 : 
        break
# every 12 months apply extra payment
   
    balance = balance - extra_yearly_payment
    print(f"{year:<4.1f} | {interest:>15,.2f} | {principal_paid:>20,.2f} | {balance:>10,.2f}")

print(f"{year:<4.1f} | {interest:>15,.2f} | {principal_paid:>20,.2f} | {balance:>10,.2f}")
print("-" * 65)
print(f"Loan will be finished in {year} year {month} months")
print("-" * 65)
````
