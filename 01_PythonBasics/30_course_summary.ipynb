## Part 4: Advanced Python Techniques (Notebook 17)  

### 🔹 List Comprehensions  
- Concise way to create lists: `[expression for item in iterable if condition]`  

```python
# Traditional loop
squares = []
for x in range(5):
    squares.append(x**2)

# List comprehension
squares = [x**2 for x in range(5)]
```

### 🔹 Generators  
- Produce items one at a time using `yield`, saving memory.  

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)   # 1 2 3
```

## Part 7: Object‑Oriented Programming (Notebooks 23–29)  

### 🔹 Classes and Instances (23)  
- A **class** is a blueprint; an **instance** is an object created from it.  
- Attributes store data; methods define behaviour.  

```python
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Rahul")
print(s.name)
```

### 🔹 State and Behaviour (24)  
- **State** = attributes (e.g., `name`, `marks`).  
- **Behaviour** = methods (e.g., `total_marks()`).  
- Methods can modify an object’s internal state.  

### 🔹 Encapsulation (25)  
- Restrict direct access to data using **private** (`__var`) and **protected** (`_var`) attributes.  
- Provide getters/setters with validation.  

```python
class Student:
    def __init__(self, marks):
        self.__marks = marks

    def get_marks(self):
        return self.__marks.copy()
```

### 🔹 Inheritance (26)  
- Create a child class that inherits from a parent.  
- Reuse code and add specialised behaviour.  

```python
class ScienceStudent(Student):
    def __init__(self, name, science_marks):
        super().__init__(name)
        self.science_marks = science_marks
```

### 🔹 Polymorphism (27)  
- Same method name, different implementations.  
- Achieved through method overriding and duck typing.  

```python
def print_info(person):
    person.display_info()   # works for Student, Teacher, etc.
```

### 🔹 Advanced Class Methods & Static Methods (29)  
- **Instance methods** – work with object data.  
- **Class methods** (`@classmethod`) – work with class‑level data.  
- **Static methods** (`@staticmethod`) – utility functions related to the class.  

```python
class Student:
    total = 0
    def __init__(self, name):
        self.name = name
        Student.total += 1

    @classmethod
    def total_students(cls):
        return cls.total

    @staticmethod
    def is_valid_mark(mark):
        return 0 <= mark <= 100
```
