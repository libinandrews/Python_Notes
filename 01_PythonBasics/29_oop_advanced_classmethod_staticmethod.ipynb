# 29. Advanced Class Methods and Static Methods  
## Taking OOP to the Next Level  

Welcome! In this notebook, we’ll explore three types of methods you can define in a Python class:  

- **Instance methods** – the ones we’ve been using (they work with object data).  
- **Class methods** – methods that work with class‑level data (shared by all instances).  
- **Static methods** – utility methods that belong to the class logically but don’t need access to instance or class data.  

Understanding these will help you design cleaner, more organised programs.  

We’ll continue with our **Kerala SSLC mark sheet** example to make everything concrete.

---

## 1. Quick Recap – Instance Methods  

An **instance method** is the most common type. It takes `self` as the first parameter and can access and modify the object’s attributes.  


```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks   # list of marks

    def total_marks(self):          # instance method
        return sum(self.marks)

    def display(self):               # instance method
        print(f"{self.name}: {self.marks} → Total: {self.total_marks()}")

# Usage
s1 = Student("Rahul", [95, 88, 79, 91])
s1.display()
```


Instance methods are tied to a specific object. They can read and modify `self`’s attributes.

---

## 2. Class Attributes vs Instance Attributes  

Before we talk about class methods, we need to understand **class attributes**.  
A class attribute is shared by all instances of the class.  


```python
class Student:
    # Class attribute
    school_name = "Kerala Public School"

    def __init__(self, name):
        self.name = name        # instance attribute

# Access class attribute via class or instance
print(Student.school_name)

s1 = Student("Rahul")
s2 = Student("Lakshmi")
print(s1.school_name)   # same for all instances
print(s2.school_name)

# Changing class attribute affects all instances
Student.school_name = "GHSS Trivandrum"
print(s1.school_name)
print(s2.school_name)
```


Class attributes are useful for data that should be the same across all objects (like school name, passing marks, etc.).

---

## 3. Class Methods – Working with Class‑Level Data  

A **class method** is defined using the `@classmethod` decorator. Its first parameter is `cls` (the class itself), not `self`.  
Class methods can access and modify class attributes, but they cannot access instance‑specific data (because they don’t have a `self`).

### Example: Keeping Track of Total Students  

Let’s use a class attribute `total_students` and a class method to increment it.

```python

class Student:
    total_students = 0          # class attribute

    def __init__(self, name):
        self.name = name
        Student.total_students += 1   # increment when a new student is created

    @classmethod
    def get_total(cls):
        """Class method to return the total number of students."""
        return cls.total_students

# Create students
s1 = Student("Rahul")
s2 = Student("Lakshmi")
s3 = Student("Arjun")

# Call class method – note we call it on the class, not on an instance
print(f"Total students: {Student.get_total()}")

# We can also call it on an instance, but it's better to use the class
print(s1.get_total())   # works, but not recommended
```


The class method `get_total()` accesses the class attribute `total_students` via `cls`. This is cleaner than hardcoding `Student.total_students` inside the method.

## 4. Class Methods as Factory Methods  

A common use of class methods is to provide alternative constructors – often called **factory methods**.  
For example, we might want to create a `Student` object from a string like `"Rahul,95,88,79,91"`.

```python

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"{self.name}: {self.marks}")

    @classmethod
    def from_string(cls, data_string):
        """Create a Student from a comma‑separated string: name,mark1,mark2,..."""
        parts = data_string.split(',')
        name = parts[0]
        marks = [int(m) for m in parts[1:]]   # convert marks to integers
        return cls(name, marks)                # call the class constructor

# Usage
data = "Rahul,95,88,79,91"
s = Student.from_string(data)
s.display()
```


The class method `from_string` acts as an alternative constructor. It parses the string and returns a new `Student` instance.

## 5. Static Methods – Utility Functions  

A **static method** is defined using the `@staticmethod` decorator. It does **not** receive `self` or `cls`.  
Static methods are just like regular functions, but they belong to the class’s namespace. They are used for utility tasks that are related to the class but don’t need access to any instance or class data.

### Example: Validating Marks  

We might want a method that checks if a given mark is valid (between 0 and 100). This doesn’t need any object or class data – it’s a pure function.

```python

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    @staticmethod
    def is_valid_mark(mark):
        """Return True if mark is between 0 and 100."""
        return 0 <= mark <= 100

    def add_mark(self, new_mark):
        if Student.is_valid_mark(new_mark):
            self.marks.append(new_mark)
        else:
            print(f"Invalid mark: {new_mark}")

# Use static method without creating an object
print(Student.is_valid_mark(85))   # True
print(Student.is_valid_mark(105))  # False

# Use it inside an instance method
s = Student("Rahul", [95, 88])
s.add_mark(79)
s.add_mark(110)   # invalid
print(s.marks)
```


Static methods are often used for validation, formatting, or any helper function that is conceptually tied to the class.

## 6. Combining All Three in a Realistic Example  

Let’s build a more comprehensive `Student` class that uses:

- **Instance methods** to calculate total, percentage, grade.  
- **Class attributes** for passing marks and school name.  
- **Class methods** to track total students and create students from a file‑like string.  
- **Static methods** to validate marks and calculate grade boundaries.


```python

class Student:
    # Class attributes
    school_name = "Kerala Public School"
    passing_marks = 40   # minimum marks to pass a subject
    total_students = 0

    def __init__(self, name, marks_dict):
        self.name = name
        self.marks = marks_dict          # dictionary: subject -> mark
        Student.total_students += 1

    # Instance methods
    def total_marks(self):
        return sum(self.marks.values())

    def percentage(self):
        total_possible = len(self.marks) * 100
        return (self.total_marks() / total_possible) * 100

    def grade(self):
        per = self.percentage()
        if per >= 90:   return "A+"
        elif per >= 80: return "A"
        elif per >= 70: return "B+"
        elif per >= 60: return "B"
        elif per >= 50: return "C+"
        elif per >= 40: return "C"
        else:           return "D"

    def display(self):
        print(f"Name   : {self.name}")
        print(f"School : {Student.school_name}")
        for sub, mark in self.marks.items():
            print(f"  {sub}: {mark}")
        print(f"Total  : {self.total_marks()}")
        print(f"Percent: {self.percentage():.2f}%")
        print(f"Grade  : {self.grade()}")

    # Class method – alternative constructor from CSV string
    @classmethod
    def from_csv(cls, csv_string):
        """Format: name,subject1=mark,subject2=mark,..."""
        parts = csv_string.split(',')
        name = parts[0]
        marks_dict = {}
        for item in parts[1:]:
            sub, mark = item.split('=')
            marks_dict[sub] = int(mark)
        return cls(name, marks_dict)

    # Class method – get total students
    @classmethod
    def get_total_students(cls):
        return cls.total_students

    # Static method – validate a single mark
    @staticmethod
    def validate_mark(mark):
        return isinstance(mark, (int, float)) and 0 <= mark <= 100

    # Static method – check if a mark is passing
    @staticmethod
    def is_passing(mark):
        return mark >= Student.passing_marks   # can access class attribute via class name
```

Now let’s test all these features.

```python
# Create a student normally
marks_rahul = {
    "Malayalam": 92,
    "English": 91,
    "Hindi": 88,
    "Mathematics": 95,
    "Science": 88,
    "Social Science": 79
}
s1 = Student("Rahul", marks_rahul)
s1.display()

print("\n" + "-"*30 + "\n")

# Create a student using class method from CSV
csv_data = "Lakshmi,Malayalam=89,English=92,Hindi=85,Mathematics=96,Science=91,Social Science=88"
s2 = Student.from_csv(csv_data)
s2.display()

print("\n" + "-"*30 + "\n")

# Use static methods
print(f"Is 75 valid? {Student.validate_mark(75)}")
print(f"Is 105 valid? {Student.validate_mark(105)}")
print(f"Is 35 passing? {Student.is_passing(35)}")

print(f"Total students created: {Student.get_total_students()}")
```

## 7. When to Use Each Type of Method  

| Method Type   | First Parameter | Access to Instance Data? | Access to Class Data? | Use Case |
|---------------|-----------------|--------------------------|------------------------|----------|
| Instance method | `self`         | ✅ Yes                   | ✅ Yes (via `self.__class__` or class name) | Regular behaviour that depends on object state |
| Class method  | `cls`           | ❌ No                    | ✅ Yes                  | Factory methods, methods that need class‑level data |
| Static method | none            | ❌ No                    | ❌ No (but can access class via class name if needed) | Utility functions related to the class |

> 🧠 **Guideline**:  
> - If your method needs to access or modify the object’s attributes, it should be an **instance method**.  
> - If your method needs to work with class‑level data (like counting instances or providing alternative constructors), use a **class method**.  
> - If your method is just a helper function that doesn’t need any object or class context, make it a **static method**.

## 8. Variation – Using Class Methods to Modify Class Attributes  

Class methods can also be used to change class attributes in a controlled way. For example, we might want to change the school name for all students at once.

```python

class Student:
    school_name = "Kerala Public School"

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name
        print(f"School name changed to: {cls.school_name}")

# Before change
s1 = Student("Rahul")
print(s1.school_name)

# Change via class method
Student.change_school("GHSS Trivandrum")

# All instances reflect the change
s2 = Student("Lakshmi")
print(s1.school_name)
print(s2.school_name)
```

## 9. Summary  

- **Instance methods** are the default; they work with object data via `self`.  
- **Class methods** (decorated with `@classmethod`) take `cls` and work with class‑level data. They are often used as factory methods or to access/modify class attributes.  
- **Static methods** (decorated with `@staticmethod`) are like regular functions but belong to the class namespace. They are used for utility tasks that don’t depend on object or class state.  
- Using the right type of method makes your code more **organised, readable, and maintainable**.
