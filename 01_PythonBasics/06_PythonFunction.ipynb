# DAY 07 : Python Functions


Functions are reusable blocks of code that perform a specific task. They help organize code, avoid repetition, and make programs easier to read and debug.

Think of a function like a **mini-program** inside your main program. You give it some inputs (parameters), it does something, and optionally returns an output. We'll use your **Kerala SSLC mark sheet** as our running example.

## 1️⃣ Defining and Calling Functions

````python
def function_name(parameters):
    """Optional docstring explaining what function does"""
    # function body
    return value   # optional
````

````python
# Example: Simple function with no parameters
def show_passed_message():
    print("Congratulations! You passed the exam.")
````

````python
# Call the function
show_passed_message()
````

## 2️⃣ Functions with Parameters

````python
# Example: Check if a subject is passed
def check_pass(score, pass_mark=40):
    if score >= pass_mark:
        return True
    else:
        return False
````

````python
# Using the function
math_score = 71
if check_pass(math_score):
    print("Math: PASSED")
else:
    print("Math: FAILED")
````

````python
# Multiple Parameters
def calculate_grade(score, subject_name):
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B+"
    elif score >= 60:
        grade = "B"
    elif score >= 50:
        grade = "C+"
    elif score >= 40:
        grade = "C"
    else:
        grade = "D"
    
    print(f"{subject_name}: {score}% -> Grade {grade}")
````

````python
# Call with different subjects
calculate_grade(75, "Malayalam")
calculate_grade(82, "Chemistry")
````

## 3️⃣ Return Values
Functions can return data using the `return` statement. You can capture that data in a variable.

````python
# Example: Calculate total marks
def calculate_total(scores):
    total = 0
    for score in scores:
        total += score
    return total
````

````python
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
total_marks = calculate_total(scores)
print(f"Total marks: {total_marks}")
````

````python
# Returning Multiple Values
def get_student_info():
    name = "LAALU"
    roll = "006615"
    dob = "22/07/1985"
    return name, roll, dob   # Returns a tuple

student_name, student_roll, student_dob = get_student_info()
print(f"Name: {student_name}, Roll: {student_roll}, DOB: {student_dob}")
````

## 4️⃣ Default Parameter Values

You can assign default values to parameters. If the caller doesn't provide that argument, the default is used.

````python
def get_grade_description(grade, full_form=True):
    if full_form:
        descriptions = {
            "A+": "A Plus",
            "A": "A Only",
            "B+": "B Plus",
            "B": "Good",
            "C+": "Fair",
            "C": "Average",
            "D": "Failed"
        }
    else:
        descriptions = {
            "A+": "A+",
            "A": "A",
            "B+": "B+",
            "B": "B",
            "C+": "C+",
            "C": "C",
            "D": "D"
        }
    return descriptions.get(grade, "Unknown")
````

````python
print(get_grade_description("B+"))               # Uses default full_form=True
print(get_grade_description("A", full_form=False))
````

## 5️⃣ Keyword Arguments
You can call functions by specifying parameter names; order doesn't matter.

````python
def print_result(name, status, year):
    print(f"{name} - {status} in {year}")

# Positional arguments
print_result("LAALU", "PASSED", 2021)

# Keyword arguments (order can be changed)
print_result(year=2021, name="LAALU", status="PASSED")
````

## 6️⃣ Variable Scope

Variables defined inside a function are **local** to that function. Variables defined outside are **global**.

````python
global_school = "DESAM GOVERNMENT HIGH SCHOOL"   # Global variable

def print_school():
    local_var = "Inside function"   # Local variable
    print("School from function:", global_school)   # Can access global
    print(local_var)

print_school()
# print(local_var)   # ❌ Error: local_var not accessible outside
````

````python
# Modifying Global Variables
# To modify a global variable inside a function, use the `global` keyword.

total_students = 0

def add_student():
    global total_students
    total_students += 1
    print(f"Total students now: {total_students}")

add_student()
add_student()
````

## 7️⃣ Commonly Used Built-in Functions

Python comes with many useful built-in functions. Here are some demonstrated with mark sheet data.
### `len()` – Get length
```python
subjects = ["Malayalam", "English", "Hindi", "Science", "Maths"]
print("Number of subjects:", len(subjects))
```

### `max()` and `min()` – Find highest and lowest scores
```python
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
print("Highest score:", max(scores))
print("Lowest score:", min(scores))
```

### `sum()` – Add all numbers
```python
total = sum(scores)
print("Total marks:", total)
```

### `round()` – Round a number
```python
average = total / len(scores)
print("Average:", round(average, 2))   # Round to 2 decimal places
```

### `type()` – Check data type
```python
print(type(scores))        # <class 'list'>
print(type("LAALU"))       # <class 'str'>
```

### `sorted()` – Sort a list
```python
sorted_scores = sorted(scores, reverse=True)   # descending
print("Scores sorted (highest first):", sorted_scores)
```

### `zip()` – Combine two lists
```python
subjects = ["Malayalam", "English", "Hindi", "Social Science", "Physics", "Chemistry", "Biology", "Mathematics", "IT"]
grades = ["B+", "B+", "B+", "A", "B+", "A", "B", "B+", "A"]

for subject, grade in zip(subjects, grades):
    print(f"{subject}: {grade}")
```

### `enumerate()` – Get index while looping
```python
for index, subject in enumerate(subjects):
    print(f"{index+1}. {subject}")
```

## 8️⃣ Lambda Functions (Anonymous Functions)
Lambda functions are small, one-line functions without a name. Useful for simple operations.

### Syntax
```python
lambda arguments: expression
```

````python
# Example: Simple grade checker
is_passing = lambda score: score >= 40

print(is_passing(35))   # False
print(is_passing(75))   # True
````

````python
# Using with `map()` – Apply function to all items
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
# Convert each score to "Pass" or "Fail"
results = list(map(lambda s: "Pass" if s >= 40 else "Fail", scores))
print(results)
````

````python
## Using with `filter()` – Filter items
# Get only scores above 80
above_80 = list(filter(lambda s: s > 80, scores))
print("Scores > 80:", above_80)
````

## 9️⃣ Recursive Functions
A function that calls itself. Useful for problems that can be broken into smaller similar sub-problems.

````python
# Example: Calculate total marks recursively

def sum_scores(scores_list):
    if len(scores_list) == 0:
        return 0
    else:
        return scores_list[0] + sum_scores(scores_list[1:])

scores = [75, 74, 72]
print("Sum (recursive):", sum_scores(scores))
````

## 🔟 Practical Example: Complete Mark Sheet Analysis with Functions
Let's build a set of functions to analyze your mark sheet.

````python
# Data
subjects = ["Malayalam", "English", "Hindi", "Social Science", "Physics", "Chemistry", "Biology", "Mathematics", "IT"]
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
pass_mark = 40
grade_thresholds = [90, 80, 70, 60, 50, 40]
grade_names = ["A+", "A", "B+", "B", "C+", "C", "D"]

def get_grade(score):
    """Return grade based on score"""
    for i, threshold in enumerate(grade_thresholds):
        if score >= threshold:
            return grade_names[i]
    return grade_names[-1]   # D

def analyze_subject(subject, score):
    """Print analysis for one subject"""
    grade = get_grade(score)
    status = "PASS" if score >= pass_mark else "FAIL"
    print(f"{subject:15} {score:3}% -> {grade:2} ({status})")
    return grade

def analyze_all(subjects, scores):
    """Analyze all subjects and return summary"""
    print("=" * 40)
    print("SUBJECT ANALYSIS")
    print("=" * 40)
    
    all_grades = []
    for sub, sc in zip(subjects, scores):
        grade = analyze_subject(sub, sc)
        all_grades.append(grade)
    
    # Summary
    print("-" * 40)
    total = sum(scores)
    avg = total / len(scores)
    print(f"Total Marks: {total}")
    print(f"Average: {avg:.2f}%")
    print(f"Passed all: {all(s >= pass_mark for s in scores)}")
    print(f"Distinctions (>=80): {sum(1 for s in scores if s >= 80)}")
    
    return all_grades

def get_toppers(subjects, scores, threshold=80):
    """Return list of subjects with score >= threshold"""
    toppers = []
    for sub, sc in zip(subjects, scores):
        if sc >= threshold:
            toppers.append(sub)
    return toppers

# Main execution
grades_result = analyze_all(subjects, scores)
print("\nToppers (>=80%):", get_toppers(subjects, scores))

# Save grade card as dictionary using zip
grade_card = dict(zip(subjects, grades_result))
print("\nGrade Card:", grade_card)
````

## 📋 Summary Table: Function Concepts

| Concept | Description | Example |
|:---|:---|:---|
| **Definition** | `def` keyword | `def greet(): print("Hello")` |
| **Parameters** | Inputs to function | `def add(a, b): return a+b` |
| **Return** | Sends value back | `return result` |
| **Default args** | Default values for parameters | `def power(base, exp=2):` |
| **Keyword args** | Call with param names | `func(x=1, y=2)` |
| **Scope** | Local vs global variables | `global x` to modify global |
| **Lambda** | One-line anonymous function | `lambda x: x*2` |
| **Built-in** | Predefined functions | `len()`, `max()`, `sum()` |
| **Recursion** | Function calls itself | `def fact(n): ...` |
