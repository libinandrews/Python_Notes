# 1. PEP 8 – Python Style Guide

## What is PEP 8?

PEP 8 is the **official Python style guide** that defines how Python code should be written.

PEP stands for **Python Enhancement Proposal**.

PEP 8 helps developers write code that is:

- Readable
- Consistent
- Maintainable

Good formatting makes code easier to **review and debug**.

---



# 2. Indentation

Python uses indentation to define code blocks.

PEP 8 recommends **4 spaces per indentation level**.

Example:

```python
for i in range(5):
    print(i)
````

Avoid mixing **tabs and spaces**.

---



# 3. Line Length

Maximum recommended line length:

**79 characters**

Example:

Bad:

```python
average_marks_of_students = df[["Malayalam","English","Physics","Chemistry","Mathematics"]].mean()
```

Better:

```python
subjects = [
    "Malayalam",
    "English",
    "Physics",
    "Chemistry",
    "Mathematics"
]

average_marks = df[subjects].mean()
```

---

# 4. Naming Conventions

PEP 8 recommends consistent naming.

| Type      | Style      | Example               |
| --------- | ---------- | --------------------- |
| Variables | snake_case | `total_marks`         |
| Functions | snake_case | `calculate_average()` |
| Constants | UPPER_CASE | `MAX_SCORE`           |
| Classes   | PascalCase | `StudentData`         |

Example:

```python
total_marks = df["Total"].sum()
```

---



# 5. Spaces Around Operators

Always add spaces around operators.

Bad:

```python
total=df["Physics"]+df["Chemistry"]
```

Good:

```python
total = df["Physics"] + df["Chemistry"]
```

---


# 6. Import Formatting

Imports should be placed **at the top of the file**.

Example:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import order:

1. Standard libraries
2. Third-party libraries
3. Local modules

---


# 7. Writing Comments

Comments explain **why the code exists**.

Example:

```python
# Load student dataset
df = pd.read_csv("students.csv")
```

Use comments when logic is **not obvious**.

---


# 8. Using Blank Lines

Blank lines improve readability.

Example:

```python
import pandas as pd


df = pd.read_csv("students.csv")


print(df.head())
```

---

# 9. Writing Clear Functions

Functions should do **one task only**.

Example:

```python
def calculate_average_marks(dataframe):
    return dataframe["Total"].mean()
```

---

# 10. Tools for Checking PEP 8

Some tools automatically check Python style.

Examples:

* flake8
* pylint
* black

Install flake8:

```bash
pip install flake8
```

Run style check:

```bash
flake8 script.py
```

---

# Review & Documentation

After writing code using good style practices, the next step is **reviewing and documenting the workflow**.

Good documentation helps:

* Understand data pipelines
* Debug errors
* Maintain projects
* Collaborate with teams

---

# 11. Reading Official Documentation

Official documentation is the **best source to understand a library**.

Examples:

Pandas Documentation
[https://pandas.pydata.org/docs/](https://pandas.pydata.org/docs/)

Plotly Documentation
[https://plotly.com/python/](https://plotly.com/python/)

Example using Pandas documentation:

```python
import pandas as pd

df = pd.read_csv("students.csv")

df.head()
```

From documentation we learn functions such as:

* `read_csv()`
* `head()`
* `describe()`
* `groupby()`

---
