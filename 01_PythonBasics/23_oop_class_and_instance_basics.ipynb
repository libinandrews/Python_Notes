# Classes and Instances (OOP Basics)  

## 1. What is OOP?  

- **Object-Oriented Programming** is a way of writing programs that focuses on creating **objects**.  
- An object is a bundle of data (called **attributes**) and actions (called **methods**).  
- Think of it like a real‑world object: a student has a name, roll number, marks (attributes), and can calculate total marks or percentage (methods).  

- **Class** – a blueprint for creating objects.  
- **Instance** – an actual object created from a class.  

> 🧠 *Analogy*: A class is like the blank mark sheet form. An instance is a filled‑out mark sheet for a real student.

---

## 2. Our First Class – A Simple Student  

Let’s start with the simplest possible class. A class is defined using the `class` keyword.

```python
class Student:
    pass   # pass means "do nothing" – we'll fill it later
```

Now we can create **instances** (objects) of this class.

```python
# Create two student objects
student1 = Student()
student2 = Student()

# They are different objects in memory
print(student1)
print(student2)
```

So far, our students have no data. Let's fix that.

---
## 3. Adding Attributes to an Instance  

We can attach data to an object using dot notation.

```python
student1.name = "Rahul"
student1.roll_no = 101
student1.school = "GHSS Trivandrum"

student2.name = "Lakshmi"
student2.roll_no = 102
student2.school = "GHSS Trivandrum"

print(f"{student1.name} (Roll No: {student1.roll_no})")
print(f"{student2.name} (Roll No: {student2.roll_no})")
```

> This works, but it’s tedious to add attributes one by one. Wouldn’t it be nice if every new student automatically had a name and roll number?

---

## 4. The `__init__` Method – Initializing New Objects  

The special method `__init__` is called automatically when we create an object. It lets us set up the initial attributes.

```python
class Student:
    def __init__(self, name, roll_no, school):
        self.name = name      # self refers to the current object
        self.roll_no = roll_no
        self.school = school
```

Now when we create a student, we must provide these details.

```python
student1 = Student("Rahul", 101, "GHSS Trivandrum")
student2 = Student("Lakshmi", 102, "GHSS Trivandrum")

print(f"{student1.name} - {student1.roll_no} - {student1.school}")
print(f"{student2.name} - {student2.roll_no} - {student2.school}")
```

✅ Much cleaner!
---

## 5. Adding Marks – Attributes for Subjects  

Now let’s store marks for SSLC subjects. We can pass a list of marks to `__init__`. For now, we’ll use four common subjects: Mathematics, Science, Social Science, and English.


```python
class Student:
    def __init__(self, name, roll_no, school, marks):
        self.name = name
        self.roll_no = roll_no
        self.school = school
        self.marks = marks   # marks will be a list of numbers
```

```python
# Marks in order: [Maths, Science, Social, English]
marks_rahul = [95, 88, 79, 91]
marks_lakshmi = [89, 92, 85, 96]

student1 = Student("Rahul", 101, "GHSS Trivandrum", marks_rahul)
student2 = Student("Lakshmi", 102, "GHSS Trivandrum", marks_lakshmi)

# Print each student's marks
print(f"{student1.name}'s marks: {student1.marks}")
print(f"{student2.name}'s marks: {student2.marks}")
```

---

## 6. Adding Methods – Giving Behaviour to Objects  

A **method** is a function inside a class that works with the object’s data. Let’s add a method to calculate total marks.

```python
class Student:
    def __init__(self, name, roll_no, school, marks):
        self.name = name
        self.roll_no = roll_no
        self.school = school
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)
```

```python
student1 = Student("Rahul", 101, "GHSS Trivandrum", marks_rahul)
student2 = Student("Lakshmi", 102, "GHSS Trivandrum", marks_lakshmi)


print(f"{student1.name}'s total marks: {student1.total_marks()}")
print(f"{student2.name}'s total marks: {student2.total_marks()}")
```

---

## 7. More Useful Methods – Percentage and Grade  

In Kerala SSLC, grades are given based on marks. Let’s add methods for percentage and grade.  
We’ll assume each subject is out of 100, so total possible marks = number of subjects × 100.

```python
class Student:
    def __init__(self, name, roll_no, school, marks):
        self.name = name
        self.roll_no = roll_no
        self.school = school
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def percentage(self):
        total_possible = len(self.marks) * 100   # each subject out of 100
        return (self.total_marks() / total_possible) * 100

    def grade(self):
        per = self.percentage()
        if per >= 90:
            return "A+"
        elif per >= 80:
            return "A"
        elif per >= 70:
            return "B+"
        elif per >= 60:
            return "B"
        elif per >= 50:
            return "C+"
        elif per >= 40:
            return "C"
        else:
            return "D"
```

```python
student1 = Student("Rahul", 101, "GHSS Trivandrum", marks_rahul)
student2 = Student("Lakshmi", 102, "GHSS Trivandrum", marks_lakshmi)
print(f"{student1.name} - Percentage: {student1.percentage():.2f}% - Grade: {student1.grade()}")
print(f"{student2.name} - Percentage: {student2.percentage():.2f}% - Grade: {student2.grade()}")
```

- **Class** – a blueprint for creating objects.  
- **Instance** – an object created from a class.  
- **Attributes** – data stored in an object (like `name`, `marks`).  
- **Methods** – functions inside a class that operate on the object (like `total_marks()`).  
- **`__init__`** – a special method that initialises new objects.  
- **`self`** – refers to the current instance inside the class.  
- We built a small SSLC mark sheet system using OOP, making it easy to create and manage multiple students.
