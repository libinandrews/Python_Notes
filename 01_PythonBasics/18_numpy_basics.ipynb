## 1️⃣ What is NumPy

* NumPy stands for **Numerical Python**
* It is the **core library for numerical computing in Python**
* Provides the powerful **ndarray (N-dimensional array)**
* Faster than Python lists because operations are **vectorized**
* Widely used in:

  * Data Science
  * Machine Learning
  * Scientific Computing
  * Simulations
  
More details: https://numpy.org/

```python
# import numpy
import numpy as np
```

```python
# 2️⃣ Why NumPy is Faster than Python Lists
numbers = [1,2,3,4]

squared = []
for n in numbers:
    squared.append(n**2)

print(squared)
```

```python
# NumPy version
import numpy as np

arr = np.array([1,2,3,4])

print(arr**2)
```

Advantages:

* less code
* faster execution
* vectorized operations

```python
# Creating NumPy Arrays
# From list
arr = np.array([1,2,3,4])
```

```python
# 2D array
arr = np.array([[1,2,3],
                [4,5,6]])
```

```python
# Common array generators
np.zeros((3,3))
np.ones((2,2))
np.full((3,3), 7)
np.eye(4)
```

```python
# Creating Ranges of Numbers
np.arange(0,10)
```

```python
np.linspace(2,10,5)
```

Difference

| Function | Description     |
| -------- | --------------- |
| arange   | step based      |
| linspace | equal intervals |


```python
# Array Shape and Dimensions
arr = np.array([[1,2,3],[4,5,6]])

arr.shape
arr.ndim
arr.size
```

```python
# Indexing and Slicing
arr = np.array([10,20,30,40])

arr[0]
arr[1:3]
```

```python
# 2D indexing
arr = np.array([[1,2,3],
                [4,5,6]])

arr[0,1]
```

```python
# Array Data Types
arr = np.array([1,2,3], dtype=float)

arr.dtype
```

Common types

* int32
* float64
* bool
* complex


```python
# Array Math Operations
a = np.array([1,2,3])
b = np.array([4,5,6])

a + b
a * b
a ** 2
```

```python
# NumPy Mathematical Functions
np.sqrt(a)
np.log(a)
np.exp(a)
np.mean(a)
np.std(a)
```

```python
# Reshaping Arrays
arr = np.arange(6)

arr.reshape(2,3)
```

```python
# Transpose
arr = np.array([[1,2,3],
                [4,5,6]])

arr.T
```

```python
# Broadcasting (Important Concept)
# Broadcasting allows operations on arrays of **different shapes**

a = np.array([1,2,3])
b = 10

a + b
```

```python
# Random Numbers
rng = np.random.default_rng()

rng.random(5)
rng.integers(0,10,5)
```

```python
# Performance Comparison
import time

list_data = list(range(1_000_000))
np_data = np.arange(1_000_000)
```

```python
start = time.time()
[x**2 for x in list_data]
print("List time:", time.time() - start)
```

```python
start = time.time()
np_data**2
print("NumPy time:", time.time() - start)
```

```python
sales = np.array([100,120,90,140])

print("Total:", sales.sum())
print("Average:", sales.mean())
print("Max:", sales.max())
```
