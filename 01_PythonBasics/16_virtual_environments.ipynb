## Virtual Environments

- Different Python projects may need **different versions of the same library**.
- Installing all libraries globally can create **dependency conflicts** between projects.
- A **virtual environment (`venv`)** creates an isolated Python environment for a specific project.
- Each environment has its **own Python interpreter and installed packages**.
- This allows multiple projects to use **different package versions without affecting each other**.
- Virtual environments help make projects **portable and reproducible across systems**.
- Developers usually store project dependencies in a **`requirements.txt` file**.
- The `requirements.txt` file allows others to **install the exact same packages easily** using `pip`.

## Creating and Activating a Virtual Environment

### Create a Virtual Environment
- Python provides a built-in module called **`venv`** to create virtual environments.
- Run the following command inside your project folder:

```bash
python -m venv .venv
````

* `python -m venv` → runs the **venv module**
* `.venv` → name of the virtual environment folder (commonly used name)

After running the command, Python creates a folder structure like:

```
project_folder/
│
├── .venv/
│   ├── Scripts/ (Windows)
│   ├── bin/ (Linux/Mac)
│   └── Lib/
```

This folder contains a **separate Python installation and package directory**.

---

## Activate the Virtual Environment

### Windows (PowerShell / CMD)

```bash
.venv\Scripts\activate
```

After activation you will see:

```
(.venv) C:\project_folder>
```

---

### Linux / macOS

```bash
source .venv/bin/activate
```

After activation you will see:

```
(.venv) user@machine:project_folder$
```

---

## Deactivate the Virtual Environment

To exit the virtual environment:

```bash
deactivate
```

This returns the terminal to the **global Python environment**.

```
```
