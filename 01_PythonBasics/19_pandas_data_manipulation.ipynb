## What is Pandas?

- It is built on top of **NumPy**.
- It provides easy tools to work with **structured data (tables)**.
- Pandas is widely used in:
  - Data Science
  - Machine Learning
  - Finance
  - Data Engineering
  - Research

Main features:

- Easy data loading (CSV, Excel, JSON, SQL)
- Data cleaning and transformation
- Data aggregation and analysis
- Time series support
- Integration with visualization libraries

# Core Data Structures

Pandas provides two main data structures.

## 1️⃣ Series

- A **Series** is a one-dimensional labeled array.
- Similar to a column in a table.

````python
import pandas as pd

data = pd.Series([10,20,30,40])

print(data)
````

## 2️⃣ DataFrame

* A **DataFrame** is a two-dimensional table.
* Similar to an **Excel sheet or SQL table**.
* Rows → records
* Columns → variables


````python
data = {
    "Name": ["Alice","Bob","Charlie"],
    "Age": [25,30,35],
    "City": ["NY","LA","Chicago"]
}

df = pd.DataFrame(data)

print(df)
````

# Creating DataFrames

````python
# From Dictionary
df = pd.DataFrame({
    "A":[1,2,3],
    "B":[4,5,6]
})
df
````

````python
# From List
data = [
    ["Alice",25],
    ["Bob",30]
]

df = pd.DataFrame(data, columns=["Name","Age"])
df
````
