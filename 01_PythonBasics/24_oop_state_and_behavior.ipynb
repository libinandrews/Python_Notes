# 24. Object-Oriented Programming: State and Behavior  

## Introduction

In the previous notebook, we learned about **classes** (blueprints) and **instances** (objects created from those blueprints). Now we will explore two fundamental ideas that make objects powerful:  

- **State** – the data an object holds (its attributes).  
- **Behavior** – the actions an object can perform (its methods).  

> 💡 **Think of a real-world object**: A student has a name, roll number, and marks (state). The student can calculate total marks, find percentage, and request a grade (behavior).  

By the end of this notebook, you’ll understand how state and behavior work together to model real‑world systems in code.

---

## 1. Quick Recap – Classes and Instances  

Let’s quickly create a simple `Student` class to remind ourselves of the syntax. We’ll build on this throughout the notebook.

```python
class Student:
    def __init__(self, name, roll_no):
        self.name = name      # attribute (state)
        self.roll_no = roll_no
```

```python
student1 = Student("Rahul", 101)
student2 = Student("Lakshmi", 102)

print(student1.name, student1.roll_no)
print(student2.name, student2.roll_no)
```

> Each object has its **own** copy of the attributes `name` and `roll_no`. That is its **state**.
---

## 2. What is State?  

- **State** is the collection of all attributes (variables) that belong to an object.  
- Attributes describe the current condition or properties of the object.  
- For a student, state could include `name`, `roll_no`, `marks`, `school`, etc.  

Let’s add marks to our student’s state.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks   # a list of marks – part of state
```

```python
# Now create two students with different marks:
marks_rahul = [95, 88, 79, 91]      # Maths, Science, Social, English
marks_lakshmi = [89, 92, 85, 96]

student1 = Student("Rahul", 101, marks_rahul)
student2 = Student("Lakshmi", 102, marks_lakshmi)

print(f"{student1.name}'s marks: {student1.marks}")
print(f"{student2.name}'s marks: {student2.marks}")
```

Each student now has its own **state** (name, roll number, and marks).

## 3. What is Behavior?  

- **Behavior** is what an object can **do**.  
- In code, behavior is implemented using **methods** – functions defined inside a class.  
- Methods can access and sometimes modify the object’s state.

Let’s add a method to calculate the total marks.


```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks
        
    def total_marks(self):          # method – behavior
        return sum(self.marks)       # uses the object's state (marks)
```

```python
student1 = Student("Rahul", 101, marks_rahul)
student2 = Student("Lakshmi", 102, marks_lakshmi)

print(f"{student1.name}'s total: {student1.total_marks()}")
print(f"{student2.name}'s total: {student2.total_marks()}")
```


The method `total_marks()` uses the object’s internal state (`self.marks`) to produce a result.

---

## 4. Methods Can Modify State  

So far, our methods only read the state. But methods can also **change** the state. This is important for modelling real‑world changes, like updating marks after a revaluation.

Let’s add a method that updates the mark for a specific subject.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def update_mark(self, subject_index, new_mark):
        """Update the mark for a given subject (0 = Maths, 1 = Science, etc.)"""
        if 0 <= subject_index < len(self.marks):
            old_mark = self.marks[subject_index]
            self.marks[subject_index] = new_mark
            print(f"Updated {self.name}'s mark from {old_mark} to {new_mark}")
        else:
            print("Invalid subject index!")
```

```python
student1 = Student("Rahul", 101, marks_rahul)
student2 = Student("Lakshmi", 102, marks_lakshmi)

# Now we can change a student’s state:
print("Before update:", student1.marks)
student1.update_mark(0, 98)   # update Maths (index 0) to 98
print("After update :", student1.marks)
```


The method **modified** the object’s state. This is a key idea: objects can change over time through their behavior.

---

## 5. Each Object Has Its Own State and Behavior  

It’s important to remember that each object has its **own independent state**. Behavior works on that object’s specific data.

```python
print(f"{student1.name}'s marks: {student1.marks}")
print(f"{student2.name}'s marks: {student2.marks}")

# Update only student2's Science mark (index 1)
student2.update_mark(1, 95)

print(f"{student2.name}'s marks now: {student2.marks}")
# student1's marks remain unchanged
print(f"{student1.name}'s marks still: {student1.marks}")
```


Even though both objects are of the same class, their states are separate. The `update_mark` method only affects the object it is called on.

---

## 6. Modeling Real‑World Systems  

State and behavior allow us to model real‑world systems naturally. Let’s add more realistic behavior to our student class – for example, calculating the grade based on updated marks.


```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def percentage(self):
        total_possible = len(self.marks) * 100
        return (self.total_marks() / total_possible) * 100

    def grade(self):
        per = self.percentage()
        if per >= 90:   return "A+"
        elif per >= 80: return "A"
        elif per >= 70: return "B+"
        elif per >= 60: return "B"
        elif per >= 50: return "C+"
        elif per >= 40: return "C"
        else:           return "D"

    def update_mark(self, subject_index, new_mark):
        if 0 <= subject_index < len(self.marks):
            old = self.marks[subject_index]
            self.marks[subject_index] = new_mark
            print(f"Mark updated for {self.name}: subject {subject_index} {old} → {new_mark}")
        else:
            print("Invalid index")

    def show_details(self):
        print(f"Name: {self.name}, Roll: {self.roll_no}")
        print(f"Marks: {self.marks}")
        print(f"Total: {self.total_marks()}, Percentage: {self.percentage():.2f}%, Grade: {self.grade()}")
```

Now we can simulate a real scenario: a student gets one mark increased after revaluation, and we see the grade change.

```python
# Create a student
rahul = Student("Rahul", 101, [95, 88, 79, 91])
rahul.show_details()

# Suppose Maths mark was increased by 5 after revaluation
rahul.update_mark(0, 100)
rahul.show_details()
```

## Summary  

- **State** = the attributes (data) of an object.  
- **Behavior** = the methods (actions) of an object.  
- Methods can **read** and **modify** an object’s state.  
- Each object has its own independent state; methods act on that specific object’s data.  
- By combining state and behavior, we can model real‑world entities (like a student with marks and grade calculations) in a natural, understandable way.
