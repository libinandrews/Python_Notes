## Project Overview  

Our school management system will have:

- **Person** – a base class for anyone in the school (students, teachers).  
- **Student** – inherits from Person, adds marks, roll number, stream, and grade calculation.  
- **Teacher** – inherits from Person, adds subject, salary, and experience.  
- **Course** – represents a subject taught by a teacher.  
- **School** – a container that holds lists of students and teachers, and provides methods to manage them.  

We’ll use:

- **Inheritance** – Student and Teacher are kinds of Person.  
- **Encapsulation** – protect sensitive data like marks and salary with private attributes and controlled access.  
- **Polymorphism** – the same method name (`display_info`) works differently for students and teachers.  
- **Composition** – a School *has* students and teachers.

## Step 1: Creating the Base Class `Person` (with Encapsulation)  

The `Person` class will store common attributes like name, age, and address.  
We’ll make `__name` private to demonstrate encapsulation.

```python

class Person:
    def __init__(self, name, age, address):
        self.__name = name       # private attribute
        self.__age = age         # private
        self.__address = address # private

    # Getter methods for private attributes
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_address(self):
        return self.__address

    # Setter methods with validation
    def set_name(self, new_name):
        if isinstance(new_name, str) and new_name.strip():
            self.__name = new_name
        else:
            print("Invalid name!")

    def set_age(self, new_age):
        if isinstance(new_age, (int, float)) and new_age > 0:
            self.__age = new_age
        else:
            print("Age must be a positive number.")

    def set_address(self, new_address):
        if isinstance(new_address, str):
            self.__address = new_address
        else:
            print("Invalid address.")

    # A method to display basic info (will be overridden)
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age : {self.__age}")
        print(f"Addr: {self.__address}")
```

```python
# Let’s test the `Person` class.
p = Person("Raju", 45, "Trivandrum")
p.display_info()

# Try to change name via setter
p.set_name("Rajesh")
print("After name change:")
p.display_info()

# Attempt invalid age
p.set_age(-5)   # should print error
```

## Step 2: Creating the `Student` Class (Inheritance + Encapsulation)  

The `Student` class inherits from `Person`. It adds:

- `roll_no` (public)  
- `__marks` (private) – a dictionary of subject:mark  
- `stream` (public) – e.g., "Science", "Commerce", "Humanities"  

We’ll also add methods to calculate total, percentage, grade, and update marks with validation.


```python
class Student(Person):
    def __init__(self, name, age, address, roll_no, stream, marks_dict):
        # Call parent constructor
        super().__init__(name, age, address)
        self.roll_no = roll_no
        self.stream = stream
        self.__marks = marks_dict   # private dictionary

    # Getter for marks (returns a copy to protect original)
    def get_marks(self):
        return self.__marks.copy()

    # Setter for a single subject mark with validation
    def set_mark(self, subject, new_mark):
        if subject not in self.__marks:
            print(f"Subject '{subject}' not found.")
            return
        if not isinstance(new_mark, (int, float)) or new_mark < 0 or new_mark > 100:
            print("Mark must be a number between 0 and 100.")
            return
        self.__marks[subject] = new_mark
        print(f"Updated {subject} to {new_mark}")

    # Methods for calculations
    def total_marks(self):
        return sum(self.__marks.values())

    def percentage(self):
        total_possible = len(self.__marks) * 100
        return (self.total_marks() / total_possible) * 100

    def grade(self):
        per = self.percentage()
        if per >= 90:   return "A+"
        elif per >= 80: return "A"
        elif per >= 70: return "B+"
        elif per >= 60: return "B"
        elif per >= 50: return "C+"
        elif per >= 40: return "C"
        else:           return "D"

    # Override display_info to show student details
    def display_info(self):
        super().display_info()   # call parent to show name, age, address
        print(f"Roll No: {self.roll_no}")
        print(f"Stream : {self.stream}")
        print("Marks:")
        for sub, mark in self.__marks.items():
            print(f"  {sub}: {mark}")
        print(f"Total  : {self.total_marks()}")
        print(f"Percentage: {self.percentage():.2f}%")
        print(f"Grade  : {self.grade()}")
```

```python
marks_rahul = {
    "Malayalam": 92,
    "English": 91,
    "Hindi": 88,
    "Mathematics": 95,
    "Science": 88,
    "Social Science": 79
}
rahul = Student("Rahul", 17, "Kochi", 101, "Science", marks_rahul)
rahul.display_info()
```


Now test encapsulation: try to directly access `__marks` (will fail), and use the setter.

```python
# rahul.__marks   # Uncommenting gives AttributeError
# Update a mark via setter
rahul.set_mark("Mathematics", 98)
rahul.display_info()   # shows updated mark
```

## Step 3: Creating the `Teacher` Class (Inheritance + Encapsulation)  

The `Teacher` class also inherits from `Person`. It adds:

- `employee_id` (public)  
- `subject` (public)  
- `__salary` (private) – with getter/setter and validation  
- `experience` (public, in years)

```python
class Teacher(Person):
    def __init__(self, name, age, address, emp_id, subject, salary, experience):
        super().__init__(name, age, address)
        self.emp_id = emp_id
        self.subject = subject
        self.__salary = salary
        self.experience = experience

    # Getter for salary
    def get_salary(self):
        return self.__salary

    # Setter with validation
    def set_salary(self, new_salary):
        if isinstance(new_salary, (int, float)) and new_salary >= 0:
            self.__salary = new_salary
            print("Salary must be a non-negative number.")

    # Override display_info
    def display_info(self):
        super().display_info()
        print(f"Emp ID   : {self.emp_id}")
        print(f"Subject  : {self.subject}")
        print(f"Salary   : ₹{self.__salary}")
        print(f"Experience: {self.experience} years")
```

```python
teacher = Teacher("Ms. Lakshmi", 35, "Trivandrum", "T101", "Mathematics", 45000, 10)
teacher.display_info()
```

## Step 4: Creating the `Course` Class  

A course represents a subject taught by a teacher. This class will demonstrate **composition** – a course has a teacher.

```python
class Course:
    def __init__(self, course_name, teacher, max_students=30):
        self.course_name = course_name
        self.teacher = teacher        # Teacher object
        self.max_students = max_students
        self.enrolled_students = []   # list of Student objects

    def add_student(self, student):
        if len(self.enrolled_students) < self.max_students:
            self.enrolled_students.append(student)
            print(f"{student.get_name()} enrolled in {self.course_name}.")
        else:
            print("Course is full!")

    def display_course_info(self):
        print(f"Course: {self.course_name}")
        print(f"Teacher: {self.teacher.get_name()}")
        print(f"Enrolled Students: {len(self.enrolled_students)}/{self.max_students}")
        if self.enrolled_students:
            print("Student list:")
            for s in self.enrolled_students:
                print(f"  - {s.get_name()} (Roll: {s.roll_no})")
```

Let’s create a course and enrol some students.

```python

# Create another student
marks_lakshmi = {
    "Malayalam": 89,
    "English": 92,
    "Hindi": 85,
    "Mathematics": 96,
    "Science": 91,
    "Social Science": 88
}
lakshmi = Student("Lakshmi", 17, "Kollam", 102, "Science", marks_lakshmi)

# Create a course
math_course = Course("Mathematics", teacher, 2)

# Enrol students
math_course.add_student(rahul)
math_course.add_student(lakshmi)
math_course.add_student(Student("Arjun", 17, "Alappuzha", 103, "Commerce", {}))  # will exceed max
math_course.display_course_info()
```

## Step 5: Creating the `School` Class (Composition)  

The `School` class will manage lists of students and teachers, and provide methods to add, remove, and search.


```python

class School:
    def __init__(self, name):
        self.name = name
        self.__students = []   # private list
        self.__teachers = []   # private list

    def add_student(self, student):
        self.__students.append(student)
        print(f"Added student {student.get_name()} to {self.name}.")

    def add_teacher(self, teacher):
        self.__teachers.append(teacher)
        print(f"Added teacher {teacher.get_name()} to {self.name}.")

    def find_student(self, roll_no):
        for s in self.__students:
            if s.roll_no == roll_no:
                return s
        return None

    def find_teacher(self, emp_id):
        for t in self.__teachers:
            if t.emp_id == emp_id:
                return t
        return None

    def show_all_students(self):
        print(f"Students in {self.name}:")
        for s in self.__students:
            print(f"  - {s.get_name()} (Roll: {s.roll_no})")

    def show_all_teachers(self):
        print(f"Teachers in {self.name}:")
        for t in self.__teachers:
            print(f"  - {t.get_name()} (ID: {t.emp_id})")
```

Let’s build our school.

```python

# Create a school
my_school = School("Kerala Public School")

# Add students
my_school.add_student(rahul)
my_school.add_student(lakshmi)

# Add teacher
my_school.add_teacher(teacher)

# Display lists
my_school.show_all_students()
my_school.show_all_teachers()

# Find a student
found = my_school.find_student(101)
if found:
    print("\nFound student:")
    found.display_info()
```

## Step 6: Demonstrating Polymorphism  

We have two different classes (`Student` and `Teacher`) that both inherit from `Person` and override `display_info`.  
We can write a function that calls `display_info` on any `Person` object (or any object with that method) – that’s polymorphism.


```python
def print_person_info(person):
    print("--- Person Info ---")
    person.display_info()
    print("-------------------")

# Works for both Student and Teacher
print_person_info(rahul)
print_person_info(teacher)
```

We can also store different kinds of people in a list and loop through them:

```python
people = [rahul, teacher]
for p in people:
    print_person_info(p)
```


The same method name works for both, but each class provides its own implementation.

## Step 7: Adding a Variation – Different Student Types (Inheritance + Polymorphism)  

To show inheritance further, we can create specialized student classes (like `ScienceStudent`, `CommerceStudent`) that override the `grade` method or add stream‑specific behaviour.  
But since we already have a `stream` attribute, we might instead show how polymorphism works if we have separate classes.

Let’s quickly define a `CommerceStudent` subclass of `Student` that overrides the grade calculation (e.g., different grading scale).


```python
class CommerceStudent(Student):
    def grade(self):
        per = self.percentage()
        # Commerce grading might be slightly different
        if per >= 85:   return "A"
        elif per >= 70: return "B"
        elif per >= 55: return "C"
        else:           return "D"
```

```python
marks_arjun = {
    "Malayalam": 80,
    "English": 85,
    "Hindi": 78,
    "Accountancy": 91,
    "Business Studies": 85,
    "Economics": 79
}
arjun = CommerceStudent("Arjun", 17, "Kottayam", 103, "Commerce", marks_arjun)

# Display info – the grade method is from CommerceStudent, not Student
arjun.display_info()
```


## Summary of OOP Concepts Used  

- **Classes and Objects**: `Person`, `Student`, `Teacher`, `Course`, `School` are blueprints; we created instances.  
- **Inheritance**: `Student` and `Teacher` inherit from `Person`. `CommerceStudent` inherits from `Student`.  
- **Encapsulation**:  
  - Private attributes (`__name`, `__marks`, `__salary`, `__students`) with getters/setters.  
  - Validation in setters prevents invalid data.  
- **Polymorphism**:  
  - `display_info()` behaves differently for each class.  
  - The same function (`print_person_info`) works with any object that has a `display_info` method (duck typing).  
  - `grade()` is overridden in `CommerceStudent`.  
- **Composition**: A `School` contains lists of students and teachers; a `Course` has a teacher and enrolled students.
