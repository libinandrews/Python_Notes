### Python Basic Data Types 

Think of data types as different kinds of information you can store and work with in a program.
Just like a mark sheet has different kinds of information (names, numbers, dates, grades), Python has different data types for different kinds of data.

We'll use your mark sheet (No. L 006615) as our example!



#### 1. String (`str`)
This is for text. Anything you put inside quotes (`" "` or `' '`) in Python is a string.

- **What it is:** A sequence of characters (letters, numbers, symbols, spaces).
- **When to use it:** For names, addresses, words, and any text you don't need to do math with.

````python
# Strings can be created with double quotes
name_of_candidate = "LAALU"
name_of_school = "DESAM GOVERNMENT HIGH SCHOOL"
fathers_name = "WELL-DONE VASU"  # Strings can even include other quotes!

# Or with single quotes
examination_centre = 'DESAM PUBLIC LIBRARY (Temporary)'
result_status = 'PASSED'

# Numbers that you don't do math with, like roll numbers or years, are often stored as strings.
register_number = "006615"  # We won't add or subtract this, so it's a string.

print(name_of_candidate)
print(name_of_school)
print(register_number)
````

#### 2. Integer (`int`)
This is for whole numbers, positive or negative, without any decimal point.

- **What it is:** A whole number.
- **When to use it:** For counting, for IDs, for anything that represents a quantity that can't be a fraction.




````python
# The number of chances is a perfect example of an integer.
number_of_chances = 6

# Admission number, without leading zeros.
admission_number = 6

print(number_of_chances)
print(admission_number)
````

> **Important Note:** Notice how `register_number` was a string (`"006615"`) but `number_of_chances` is an integer (`6`). If you try to do addition on a string, Python will just combine (concatenate) them. If you do addition on an integer, it will give you the mathematical sum.


````python
print("006" + "615") # This is string concatenation
print(6 + 6)          # This is integer addition
````

#### 3. Float (`float`)
This is for numbers with a decimal point. It's used when you need more precision.

- **What it is:** A number with a decimal point (a floating-point number).
- **When to use it:** For measurements, percentages, averages, and any numbers that aren't whole.

**Examples from your mark sheet:**
We don't see the exact percentage on the sheet, but we know the grades are based on percentages.
```python
# Let's calculate a hypothetical average percentage from your grades.
# Imagine you got these scores in your four main subjects.
malayalam_score = 85.5
english_score = 78.0
social_science_score = 92.0
mathematics_score = 71.5

total_score = malayalam_score + english_score + social_science_score + mathematics_score
average_percentage = total_score / 4

print("Average Percentage:", average_percentage)
```
**Output:**
```
Average Percentage: 81.75
```
Here, `81.75` is a `float`.


#### 4. Boolean (`bool`)
This is the simplest data type. It represents one of two values: `True` or `False`.

- **What it is:** A truth value.
- **When to use it:** For yes/no questions, for checking conditions, for flags.


````python
passed = True  # Because your result is PASSED

# Let's check if you need to retake the exam
# Eligibility is Minimum D+ grade in each paper.
# We know you passed, so the statement "You failed any subject?" is False.
failed_any_subject = False

print("Did the candidate pass?", passed)
print("Did the candidate fail any subject?", failed_any_subject)
````

#### 5. Checking the Type
Python has a handy function called `type()` that tells you what data type a piece of data is.

````python
print(type("LAALU"))          # Output: <class 'str'>
print(type(6))                 # Output: <class 'int'>
print(type(81.75))             # Output: <class 'float'>
print(type(True))              # Output: <class 'bool'>
````
