## 02. BASIC DATA TYPES

| Type | Name | Use Case | Example |
|:---:|:---:|:---|:---|
| **String** | `str` | Text, names, words | `name = "LAALU"` |
| **Integer** | `int` | Whole numbers, counting | `marks = 85` |

### Strings (ordered, immutable sequences)
- **Indexing**: `name[0]` → first character
- **Negative indexing**: `name[-1]` → last character
- **Slicing**: `name[0:3]` → first 3 characters

```python
# Type Checking & Conversion

# Check type
print(type("LAALU"))       # <class 'str'>
print(type(85))            # <class 'int'>
print(type(81.75))         # <class 'float'>
print(type(True))          # <class 'bool'>

# Type conversion
x = 10
y = float(x)        
print(y)                   # 10.0

z = str(x)          
print(z)                   # "10"

b = bool(x)         
print(b)                   # True (non-zero = True)
```

## 03. BASIC OPERATIONS

### Arithmetic Operations
- **Addition (`+`)**: `5 + 3 = 8`
- **Subtraction (`-`)**: `5 - 3 = 2`
- **Multiplication (`*`)**: `5 * 3 = 15`
- **Division (`/`)**: `5 / 3 = 1.666...` (returns float)
- **Floor Division (`//`)**: `5 // 3 = 1` (integer result)
- **Modulus (`%`)**: `5 % 3 = 2` (remainder)
- **Exponentiation (`**`)**: `5 ** 3 = 125`

### Comparison Operations (return `bool`)
- **Equal to**: `85 == 85` → True
- **Not equal to**: `85 != 75` → True
- **Greater than**: `85 > 75` → True
- **Less than**: `85 < 95` → True
- **Greater than or equal**: `85 >= 85` → True
- **Less than or equal**: `85 <= 90` → True

### Logical Operations
- **`and`**: Both conditions must be True
- **`or`**: At least one condition must be True
- **`not`**: Reverses the condition

### Other Operations
- **Assignment**: `x = 10`, `x += 5` (shorthand: `+=`, `-=`, `*=`, `/=`, etc.)
- **Membership**: `"Math" in grades` (checks if item in sequence)
- **Identity**: `x is y` (checks if same object in memory)

```python
# Arithmetic Operations
a = 5
b = 3

print(f"{a} + {b} = {a + b}")          # Addition
print(f"{a} - {b} = {a - b}")          # Subtraction
print(f"{a} * {b} = {a * b}")          # Multiplication
print(f"{a} / {b} = {a / b}")          # Division
print(f"{a} // {b} = {a // b}")        # Floor Division
print(f"{a} % {b} = {a % b}")          # Modulus
print(f"{a} ** {b} = {a ** b}")        # Exponentiation
```

```python
# Comparison Operations
score1 = 85
score2 = 75

print(f"{score1} == {score1}: {score1 == score1}")    # Equal to
print(f"{score1} != {score2}: {score1 != score2}")    # Not equal to
print(f"{score1} > {score2}: {score1 > score2}")      # Greater than
print(f"{score1} < {score2}: {score1 < score2}")      # Less than
print(f"{score1} >= {score1}: {score1 >= score1}")    # Greater or equal
print(f"{score1} <= {score2}: {score1 <= score2}")    # Less or equal
```

```python
# Logical Operations
score = 85

# and: both conditions must be True
result1 = (score >= 80) and (score < 90)
print(f"(score >= 80) and (score < 90): {result1}")

# or: at least one condition must be True
result2 = (score < 80) or (score >= 80)
print(f"(score < 80) or (score >= 80): {result2}")

# not: reverses the condition
result3 = not (score < 80)
print(f"not (score < 80): {result3}")
```

## 04. DATA STRUCTURES

### 1. List `[ ]` – Ordered, Mutable
- Add/remove/modify items
- Allows duplicates
- Use for: dynamic collections

```python
# List Operations
subjects = ["Malayalam", "English", "Hindi", "Math", "Science"]

# Indexing
print("First subject:", subjects[0])     # "Malayalam"
print("Last subject:", subjects[-1])     # "Science"

# Slicing
print("First three:", subjects[0:3])     # ["Malayalam", "English", "Hindi"]

# Methods
subjects.append("PE")
print("After append:", subjects)

subjects.remove("PE")
print("After remove:", subjects)

print("Length:", len(subjects))
print("Is 'Math' in subjects?", "Math" in subjects)
```

### 2. Tuple `( )` – Ordered, Immutable
- Fixed, cannot change after creation
- Faster than lists
- Use for: fixed data (DOB, registration numbers)

```python
# Tuple Operations
dob = (22, "July", 1985)
exam_info = ("006615", "March", 2021)

# Access by index
day = dob[0]
month = dob[1]
year = dob[2]

print(f"Date of Birth: {day} {month} {year}")
print(f"Exam Details: {exam_info}")

# Unpacking
day, month, year = dob
print(f"Unpacked - Day: {day}, Month: {month}, Year: {year}")

# Trying to modify will cause error (uncomment to see)
# dob[0] = 23  # ❌ TypeError
```

### 3. Dictionary `{ }` – Key-Value Pairs
- Unordered (ordered by insertion in Python 3.7+)
- Keys must be unique and immutable
- Use for: data mapping

```python
# Dictionary Operations
grades = {
    "Malayalam": "B+",
    "English": "B+",
    "Chemistry": "A",
    "Mathematics": "B+"
}

# Access
print("Chemistry grade:", grades["Chemistry"])
print("Math grade (safe):", grades.get("Math", "N/A"))

# Modify/Add
grades["Physics"] = "B"
print("After adding Physics:", grades)

# Iterate
print("\nAll grades:")
for subject, grade in grades.items():
    print(f"  {subject}: {grade}")

print(f"\nTotal subjects: {len(grades)}")
print("Chemistry in grades?", "Chemistry" in grades)
```

### 4. Set `{ }` – Unordered, Unique Elements
- No duplicates
- Unordered
- Use for: removing duplicates, membership testing

```python
# Set Operations
a_subjects = {"Chemistry", "Social Science", "IT"}
bplus_subjects = {"Malayalam", "English", "Physics", "Math"}

print("A subjects:", a_subjects)
print("B+ subjects:", bplus_subjects)

# Union - all elements from both sets
all_good = a_subjects.union(bplus_subjects)
print("\nUnion (all A or B+ subjects):", all_good)

# Intersection - common elements
common = a_subjects.intersection(bplus_subjects)
print("Intersection (common subjects):", common)

# Difference - in first but not second
only_a = a_subjects.difference(bplus_subjects)
print("Only A subjects:", only_a)

# Add/Remove
unique_grades = {"A", "B+", "B"}
unique_grades.add("A+")
print("\nAfter add:", unique_grades)
unique_grades.remove("B")
print("After remove:", unique_grades)
```

## 05. CONTROL FLOW

### Conditional Statements
Use `if`, `elif`, and `else` to make decisions based on conditions.

```python
# if statement
math_score = 71
if math_score >= 40:
    print("Mathematics: PASSED")

# if-else
if math_score >= 80:
    print("Grade: A")
else:
    print("Grade: Below A")

# if-elif-else (multiple conditions)
chemistry = 82
if chemistry >= 90:
    grade = "A+"
elif chemistry >= 80:
    grade = "A"
elif chemistry >= 70:
    grade = "B+"
else:
    grade = "B"
    
print(f"Chemistry Grade: {grade}")
```

### for Loop – Iterate Over Sequences
Use `for` to loop through lists, tuples, strings, and other sequences.

```python
# Loop over list
print("Subjects:")
for subject in ["Math", "English", "Science"]:
    print(f"  - {subject}")

# Loop with index
print("\nIndex loop (0 to 4):")
for i in range(5):
    print(f"  Index: {i}")

# Unpacking in loop
subjects = ["Math", "Science"]
grades = ["A", "B+"]
print("\nSubject-Grade pairs:")
for subject, grade in zip(subjects, grades):
    print(f"  {subject}: {grade}")
```

### while Loop – Repeat While Condition True
Use `while` to repeat code as long as a condition is true.

```python
# while loop
score = 30
pass_mark = 40
attempts = 0

print("Improving score:")
while score < pass_mark:
    attempts += 1
    score += 5
    print(f"  Attempt {attempts}: Score = {score}")

print(f"Passed after {attempts} attempts!")
```

### Loop Control Statements
- **`break`**: Exit loop immediately
- **`continue`**: Skip to next iteration
- **`pass`**: Do nothing (placeholder)

```python
# break: exit loop immediately
print("Using break:")
for score in [75, 74, 32, 71]:
    if score < 40:
        print(f"  Failed! Score: {score} (BREAK)")
        break
    print(f"  Passed! Score: {score}")

# continue: skip to next iteration
print("\nUsing continue:")
for subject in ["Math", "Physics", "English", "Chemistry"]:
    if subject == "Physics" or subject == "Chemistry":
        continue  # Skip science subjects
    print(f"  {subject}")

# pass: do nothing (placeholder)
print("\nUsing pass:")
score = 85
if score >= 80:
    pass  # TODO: Add A grade handling later
else:
    print("Grade is below A")
```

### Loop else Clause
The `else` block executes only if the loop finishes normally (without `break`).

```python
# Loop else: runs if loop completes without break
scores = [75, 74, 72, 85, 71]
pass_mark = 40

print("Checking scores:")
for score in scores:
    if score < pass_mark:
        print("Found a failing subject!")
        break
    print(f"  Score {score} is passing.")
else:
    print("All subjects passed! No failures found.")
```

### Exception Handling
Use `try`, `except`, `else`, and `finally` to handle errors gracefully.

```python
# Exception handling
try:
    # Try to calculate percentage
    total_marks = 375
    num_subjects = 0  # This will cause an error
    percentage = total_marks / num_subjects
    print(f"Percentage: {percentage}%")
except ZeroDivisionError:
    print("Error: Cannot divide by zero! Number of subjects cannot be zero.")
else:
    print("Calculation successful!")  # Runs if no error
finally:
    print("Grade calculation attempt finished.")  # Always runs
```

## 06. FUNCTIONS

Functions are reusable blocks of code that perform a specific task.

### Defining and Calling Functions

```python
# Simple function with no parameters
def show_passed_message():
    """Function with no parameters"""
    print("Congratulations! You passed the exam.")

show_passed_message()

# Function with parameters
def check_pass(score, pass_mark=40):
    """Check if score qualifies as pass"""
    if score >= pass_mark:
        return True
    else:
        return False

math_score = 71
if check_pass(math_score):
    print(f"Math ({math_score}): PASSED")
else:
    print(f"Math ({math_score}): FAILED")
```

```python
# Function with multiple parameters and return value
def calculate_grade(score, subject_name):
    """Calculate grade based on score"""
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B+"
    elif score >= 60:
        grade = "B"
    elif score >= 50:
        grade = "C+"
    elif score >= 40:
        grade = "C"
    else:
        grade = "D"
    
    return f"{subject_name}: {score}% -> Grade {grade}"

# Call with different subjects
print(calculate_grade(75, "Malayalam"))
print(calculate_grade(82, "Chemistry"))
```

### Return Values
Functions can return single or multiple values.

```python
# Single return value
def calculate_total(scores):
    """Calculate total marks"""
    return sum(scores)

scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
total_marks = calculate_total(scores)
print(f"Total marks: {total_marks}")

# Multiple return values (as tuple)
def get_student_info():
    """Return multiple values"""
    name = "LAALU"
    roll = "006615"
    dob = "22/07/1985"
    return name, roll, dob

student_name, student_roll, student_dob = get_student_info()
print(f"Name: {student_name}, Roll: {student_roll}, DOB: {student_dob}")
```

### Default Parameters & Keyword Arguments

```python
# Default parameter values
def get_grade_description(grade, full_form=True):
    """Get grade description"""
    if full_form:
        descriptions = {
            "A+": "A Plus",
            "A": "A Only",
            "B+": "B Plus",
            "B": "Good",
            "C+": "Fair"
        }
    else:
        descriptions = {
            "A+": "A+",
            "A": "A",
            "B+": "B+",
            "B": "B",
            "C+": "C+"
        }
    return descriptions.get(grade, "Unknown")

print(get_grade_description("B+"))               # Uses default full_form=True
print(get_grade_description("A", full_form=False))  # Override default

# Keyword arguments (order doesn't matter)
def print_result(name, status, year):
    print(f"{name} - {status} in {year}")

print_result("LAALU", "PASSED", 2021)  # Positional
print_result(year=2021, name="LAALU", status="PASSED")  # Keyword
```

### Variable Arguments (*args, **kwargs)

```python
# *args: variable number of positional arguments
def sum_scores(*scores):
    """Sum any number of scores"""
    return sum(scores)

print(f"Sum of 75, 80, 85: {sum_scores(75, 80, 85)}")
print(f"Sum of 60, 70, 80, 90: {sum_scores(60, 70, 80, 90)}")

# **kwargs: keyword arguments
def print_student(**info):
    """Print student info from keyword arguments"""
    for key, value in info.items():
        print(f"  {key}: {value}")

print("\nStudent Information:")
print_student(name="LAALU", roll="006615", dob="22/07/1985", status="PASSED")
```

### Variable Scope
- **Local variables**: defined inside function, only accessible inside
- **Global variables**: defined outside function, accessible everywhere
- Use `global` keyword to modify global variables inside a function

```python
# Global variable
global_school = "DESAM GOVERNMENT HIGH SCHOOL"

def print_school():
    """Function accessing global variable"""
    local_var = "Inside function"
    print(f"School from function: {global_school}")
    print(f"Local variable: {local_var}")

print_school()
# print(local_var)  # ❌ Error: not accessible outside

# Modifying global variable
total_students = 0

def add_student():
    """Modify global variable inside function"""
    global total_students
    total_students += 1
    print(f"Total students now: {total_students}")

print("\nAdding students:")
add_student()
add_student()
add_student()
```

### Lambda Functions (One-Line Anonymous Functions)
Lambda functions are small, one-line functions without a name.

```python
# Simple lambda
is_passing = lambda score: score >= 40
print(f"Is 35 passing? {is_passing(35)}")
print(f"Is 75 passing? {is_passing(75)}")

# Using with map() - Apply function to all items
scores = [75, 74, 32, 85, 38, 92]
results = list(map(lambda s: "Pass" if s >= 40 else "Fail", scores))
print(f"\nScores: {scores}")
print(f"Results: {results}")

# Using with filter() - Keep only items that satisfy condition
above_80 = list(filter(lambda s: s > 80, scores))
print(f"\nScores above 80: {above_80}")
```

### Built-in Functions
Python provides many useful built-in functions for common tasks.

```python
# Built-in functions demo
subjects = ["Malayalam", "English", "Hindi", "Social Science", "Physics", "Chemistry", "Biology", "Mathematics", "IT"]
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]

print(f"len(): {len(subjects)} subjects")
print(f"max(): Highest score is {max(scores)}")
print(f"min(): Lowest score is {min(scores)}")
print(f"sum(): Total marks: {sum(scores)}")
print(f"round(): Average rounded: {round(sum(scores)/len(scores), 2)}")
print(f"sorted(): Sorted scores: {sorted(scores)[:3]} (first 3)")

# enumerate() - Get index while looping
print("\nenumerate() - First 3 subjects with index:")
for index, subject in enumerate(subjects[:3]):
    print(f"  {index+1}. {subject}")

# zip() - Combine two lists
print("\nzip() - Subjects with scores:")
for subject, score in zip(subjects[:3], scores[:3]):
    print(f"  {subject}: {score}")
```
