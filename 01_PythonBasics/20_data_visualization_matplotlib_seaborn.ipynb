

# Data Visualization in Python

## What is Data Visualization?

Data Visualization is the process of converting **raw data into visual representations** such as charts and graphs.

Instead of reading numbers from tables, visualization helps us **quickly understand insights from data**.

For example, in our **student marks dataset**, instead of checking marks one by one, we can create charts to see:

* Which subject has the highest average marks
* Distribution of student percentages
* Relationship between subjects

---

# Why Data Visualization is Important

Data visualization helps us to:

* Understand large datasets easily
* Identify **patterns and trends**
* Detect **outliers or unusual values**
* Compare different categories
* Present results clearly in reports

Example using our dataset:

* Compare **average marks of subjects**
* See **how marks are distributed among students**

# Student Performance Analysis using Data Visualization
In this section we will analyze the **student marks dataset** using **Matplotlib and Seaborn**.

The goal is to answer questions such as:

* Which subject has the **highest average score**?
* How are **student percentages distributed**?
* Do students who score well in **Physics also score well in Mathematics**?
* Which subject shows **highest variation in marks**?

First we load the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_name = r"db\demo_student_marksheet_dataset.csv"
df = pd.read_csv(file_name)

df.head()
df.columns
```

```python
# Average Marks per Subject
subjects = ["Malayalam","English","Hindi","Physics","Chemistry","Biology","Mathematics"]

avg_marks = df[subjects].mean()

plt.figure(figsize=(8,5))
plt.bar(subjects, avg_marks)

plt.title("Average Marks per Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.xticks(rotation=45)

plt.show()
```


From this chart we can identify:

* The **highest scoring subject**
* The **lowest scoring subject**
* Overall **subject difficulty level**


```python
# Distribution of Student Percentage
plt.figure(figsize=(8,5))

sns.histplot(df["Percentage"], bins=20, kde=True)

plt.title("Distribution of Student Percentage")
plt.xlabel("Percentage")
plt.ylabel("Number of Students")

plt.show()
```

This visualization helps us see:

* Most common score range
* Whether scores are **normally distributed**
* Whether there are **very high or low performers**

```python
# Relationship Between Physics and Mathematics
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x="Percentage", y="Mathematics")

plt.title("Physics vs Mathematics Marks")

plt.show()
```

This plot helps us identify:

* Correlation between subjects
* Clusters of high-performing students
* Outliers

```python
# Subject Performance Comparison
plt.figure(figsize=(8,5))

sns.boxplot(data=df[["Physics","Chemistry","Mathematics"]])

plt.title("Distribution of Science Subject Marks")

plt.show()
```


Box plots show:

* Median marks
* Spread of marks
* Outliers
* Subject difficulty

```python
# Correlation Between Subjects
plt.figure(figsize=(8,6))

correlation = df[subjects].corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Subject Correlation Heatmap")

plt.show()
```

This helps us identify:

* Subjects that are **strongly related**
* Subjects that are **independent**
* Overall student performance patterns
