# Day 6 **Python Control Flow Statements**

In **Python**, *control flow statements* determine the **order in which code executes**. They allow a program to make decisions, repeat actions, or change the execution path based on conditions.

# 1️⃣ Conditional Statements (Decision Making)
These statements allow the program to **execute code based on conditions**.

```python
# If Executes a block if the condition is True.
math_score = 71
pass_mark = 40

if math_score >= pass_mark:
    print("Mathematics: PASSED")
```

```python
# if-else Runs one block if the condition is true, otherwise another block.
biology = 64
pass_mark = 40

if biology >= pass_mark:
    print("Biology: PASSED")
else:
    print("Biology: FAILED")
```

```python
# if-elif-else Used when multiple conditions must be checked.
chemistry = 82

if chemistry >= 90:
    grade = "A+"
elif chemistry >= 80:
    grade = "A"
elif chemistry >= 70:
    grade = "B+"
elif chemistry >= 60:
    grade = "B"
elif chemistry >= 50:
    grade = "C+"
elif chemistry >= 40:
    grade = "C"
else:
    grade = "D"

print(f"Chemistry Grade: {grade}")
```

# 2️⃣ Looping Statements (Iteration)

Used to **repeat a block of code multiple times**.


```python
# for :- Used for iterating over sequences (list, tuple, string, etc).
for subject in ["Malayalam", "English", "Hindi", "Maths", "Science"]:
    print("Subject:", subject)
```

```python
# while :- Repeats as long as the condition remains True. 
score = 30
pass_mark = 40
attempts = 0

while score < pass_mark:
    attempts += 1
    score += 5
    print(f"Attempt {attempts}: Score = {score}")

print(f"Passed after {attempts} attempts!")
```

# 3️⃣ Loop Control Statements

Used to **modify loop execution**.

```python
# break : Stops the loop immediately.
scores = [75, 74, 32, 71, 82]  # Hindi has 32 (fail)
pass_mark = 40

for score in scores:
    if score < pass_mark:
        print("Found failing subject! Stopping check.")
        break
    print(f"Score {score} is passing.")
```

```python
# continue Skips the current iteration and moves to the next one.
subjects = ["Malayalam", "Physics", "English", "Chemistry", "Maths"]
for subject in subjects:
    if subject == "Physics" or subject == "Chemistry":
        continue  # Skip science subjects
    print("Non-science subject:", subject)
```

```python
# pass Does nothing (placeholder). Useful when a statement is syntactically required but no action is needed.
score = 85

if score >= 80:
    pass  # TODO: Add A grade handling later
else:
    print("Grade is below A")

print("Continuing with rest of code...")
```

# 4️⃣ Loop `else` Clause

Python loops can have an `else`. Runs when the loop **finishes normally (no break)**.

```python
scores = [75, 74, 72, 85, 71]
pass_mark = 40

for score in scores:
    if score < pass_mark:
        print("Found a failing subject!")
        break
    print(f"Score {score} is passing.")
else:
    print("All subjects passed! No failures found.")
```

# 5️⃣ Exception Control Flow

Controls program flow when **errors occur**.

### `try`
### `except`
### `else`
### `finally`


```python
try:
    # Try to calculate percentage
    total_marks = 375
    num_subjects = 0  # This will cause an error
    percentage = total_marks / num_subjects
    print(f"Percentage: {percentage}%")
except ZeroDivisionError:
    print("Error: Cannot divide by zero! Number of subjects cannot be zero.")
else:
    print("Calculation successful!")  # Runs if no error
finally:
    print("Grade calculation attempt finished.")  # Always runs
```

# 6️⃣ Real Mark Sheet Example - Combining Everything

```python
# Your mark sheet data
subjects = ["Malayalam", "English", "Hindi", "Social Science", "Physics", "Chemistry", "Biology", "Mathematics", "IT"]
scores = [75, 74, 72, 85, 74, 82, 64, 71, 88]
pass_mark = 40
distinction = 80

print("=== KERALA SSLC RESULT ANALYSIS ===\n")

# Check each subject
for i in range(len(subjects)):
    subject = subjects[i]
    score = scores[i]
    
    if score >= pass_mark:
        status = "PASS"
        if score >= distinction:
            remark = "DISTINCTION"
        elif score >= 60:
            remark = "GOOD"
        else:
            remark = "SATISFACTORY"
    else:
        status = "FAIL"
        remark = "NEEDS IMPROVEMENT"
        # Uncomment next line to stop at first failure
        # break
    
    print(f"{subject:15} {score:3}% - {status:4} ({remark})")

# Count distinctions
distinction_count = 0
for score in scores:
    if score >= distinction:
        distinction_count += 1

print(f"\nTotal Distinctions: {distinction_count}")

# Check if all passed
all_passed = True
for score in scores:
    if score < pass_mark:
        all_passed = False
        break

if all_passed:
    print("RESULT: PASSED IN ALL SUBJECTS ✓")
else:
    print("RESULT: HAS FAILURES ✗")
```

# 📋 Complete List of Python Control Flow Statements

| Category | Statements |
|:---|:---|
| **Conditional** | `if`, `elif`, `else` |
| **Looping** | `for`, `while` |
| **Loop Control** | `break`, `continue`, `pass` |
| **Loop Clause** | `else` (with loops) |

# ✅ Quick Reference Table

| Statement | Purpose | Example from Mark Sheet |
|:---|:---|:---|
| `if` | Execute if True | `if score >= 40: print("Pass")` |
| `elif` | Check another condition | `elif score >= 80: print("A")` |
| `else` | Execute if all false | `else: print("Fail")` |
| `for` | Loop over sequence | `for subject in subjects:` |
| `while` | Loop while condition true | `while score < 40: score += 5` |
| `break` | Exit loop immediately | `if score < 40: break` |
| `continue` | Skip to next iteration | `if subject == "Physics": continue` |
| `pass` | Do nothing (placeholder) | `if score >= 80: pass # TODO` |
| `try` | Attempt code that may error | `try: percentage = total/num` |
| `except` | Handle specific error | `except ZeroDivisionError:` |
| `finally` | Always execute | `finally: print("Done")` |

# ⚠️ Important Notes

- **Indentation matters** – blocks are defined by indentation (4 spaces recommended)
- `=` is **assignment** (`score = 85`)
- `==` is **comparison** (`score == 85`)
- `break` exits the **entire loop**
- `continue` skips **only current iteration**
- `pass` is a **placeholder** that does nothing
- Loop `else` runs **only if no `break` occurred**
- `finally` always runs, **even if there's an error**
