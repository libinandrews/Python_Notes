## List Comprehensions and Generators

### Key Ideas

- Python provides tools to write **short and efficient loops**.
- **List comprehensions** allow creating lists using a **compact syntax**.
- They replace longer loops with **clean one-line expressions**.
- **Generators** produce values **one at a time** instead of storing everything in memory.
- Generators are useful when working with **large datasets or streams of data**.
- They help **reduce memory usage and improve performance**.

# List Comprehension

### Basic Syntax

```python
[expression for item in iterable]
````

### Syntax with Condition

```python
[expression for item in iterable if condition]
```

---

`````python
### Example: Normal Loop
numbers = [1,2,3,4,5]

squares = []
for n in numbers:
    squares.append(n**2)

print(squares)
`````

`````python
### Example: List Comprehension
squares = [n**2 for n in numbers]
print(squares)
`````

`````python
### Example: With Condition
numbers = range(10)
even_numbers = [n for n in numbers if n % 2 == 0]

print(even_numbers)
`````

---

# Generators
### When to Use Generators
Working with large datasets
Reading large files
Processing data streams
Performing lazy evaluation

### Generator Function Syntax

```python
def generator_name():
    yield value
```
### Generator Expression Syntax

```python
(expression for item in iterable)
```

---

`````python
### Example: Generator Function
def count_up_to(n):
    for i in range(n):
        yield i
`````

`````python
for num in count_up_to(5):
    print(num)
`````

`````python
# Example Problem (Large Data in Memory)
numbers = [n**2 for n in range(1_000_000_000)]
`````

* This tries to create a **huge list in memory**
* Python must store **1 billion numbers**
* This may cause **MemoryError** or crash the system

Memory usage becomes extremely high.

`````python
### Generator Function
numbers = (n**2 for n in range(1_000_000_000))
`````

### Summary

* **List comprehension** → shorter way to create lists
* **Generators** → memory efficient iteration
* Both help write **cleaner and faster Python code**
