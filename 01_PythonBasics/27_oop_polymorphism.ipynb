# 27. Object-Oriented Programming: Polymorphism  
## One Name, Many Forms – Making Your Code Flexible  

Welcome! In this notebook, we’ll explore **polymorphism** – a fancy word meaning “many forms”.  
Polymorphism allows different types of objects to respond to the **same method name** in their own way.  
This makes your code more flexible and easier to extend.  

We’ll continue with our student mark sheet example and see how the same method name (like `total_marks()` or `display_info()`) can behave differently for Science, Commerce, and Humanities students.

---

## 1. What is Polymorphism?  

- **Poly** = many, **morph** = form.  
- Polymorphism lets us use a single interface (e.g., a method name) with different implementations.  
- It allows objects of different classes to be treated as objects of a common superclass, especially when they share method names.  

> 🧠 **Real‑world analogy**: The word “play” means different things for a musician (plays an instrument), an actor (plays a role), and an athlete (plays a game). Same word, different behaviours.

---

## 2. Polymorphism Through Method Overriding  

We’ve already seen method overriding in the previous notebook on inheritance. That’s one form of polymorphism.  
A child class can **override** a method from its parent to provide stream‑specific behaviour.

### Example: Overriding `total_marks()` and `display_info()`

First, let’s define a parent `Student` class and two child classes: `ScienceStudent` and `CommerceStudent`.  
We’ll also add a `HumanitiesStudent` for good measure.

```python
class Student:
    def __init__(self, name, roll_no, school, common_marks):
        self.name = name
        self.roll_no = roll_no
        self.school = school
        self.common_marks = common_marks   # marks for common subjects (e.g., English, Second Language)

    def total_marks(self):
        return sum(self.common_marks)

    def percentage(self):
        total_possible = len(self.common_marks) * 100
        return (self.total_marks() / total_possible) * 100

    def display_info(self):
        print(f"Name   : {self.name}")
        print(f"Roll No: {self.roll_no}")
        print(f"School : {self.school}")
        print(f"Common Marks: {self.common_marks}")
        print(f"Total Marks : {self.total_marks()}")
        print(f"Percentage  : {self.percentage():.2f}%")

class ScienceStudent(Student):
    def __init__(self, name, roll_no, school, common_marks, science_marks):
        super().__init__(name, roll_no, school, common_marks)
        self.science_marks = science_marks   # Physics, Chemistry, Maths

    def total_marks(self):
        # Override: add science marks to total
        return super().total_marks() + sum(self.science_marks)

    def display_info(self):
        super().display_info()   # call parent to show common info
        print("Science Subjects:")
        subjects = ["Physics", "Chemistry", "Mathematics"]
        for sub, mark in zip(subjects, self.science_marks):
            print(f"  {sub}: {mark}")

class CommerceStudent(Student):
    def __init__(self, name, roll_no, school, common_marks, commerce_marks):
        super().__init__(name, roll_no, school, common_marks)
        self.commerce_marks = commerce_marks   # Accountancy, Business Studies, Economics

    def total_marks(self):
        return super().total_marks() + sum(self.commerce_marks)

    def display_info(self):
        super().display_info()
        print("Commerce Subjects:")
        subjects = ["Accountancy", "Business Studies", "Economics"]
        for sub, mark in zip(subjects, self.commerce_marks):
            print(f"  {sub}: {mark}")

class HumanitiesStudent(Student):
    def __init__(self, name, roll_no, school, common_marks, humanities_marks):
        super().__init__(name, roll_no, school, common_marks)
        self.humanities_marks = humanities_marks   # History, Geography, Political Science

    def total_marks(self):
        return super().total_marks() + sum(self.humanities_marks)

    def display_info(self):
        super().display_info()
        print("Humanities Subjects:")
        subjects = ["History", "Geography", "Political Science"]
        for sub, mark in zip(subjects, self.humanities_marks):
            print(f"  {sub}: {mark}")
```

Now create one student from each stream.

```python
# Common subjects: English (85), Malayalam (90)
science = ScienceStudent("Lakshmi", 101, "GHSS Trivandrum", [85, 90], [92, 88, 95])
commerce = CommerceStudent("Arjun", 102, "GHSS Trivandrum", [80, 88], [91, 85, 78])
humanities = HumanitiesStudent("Meera", 103, "GHSS Trivandrum", [75, 82], [88, 79, 84])
```

```python
print("=== SCIENCE STUDENT ===")
science.display_info()

print("\n=== COMMERCE STUDENT ===")
commerce.display_info()

print("\n=== HUMANITIES STUDENT ===")
humanities.display_info()
```


Each object **responds to the same method name** in its own way. That’s polymorphism in action!

## 3. Polymorphism in Functions  

We can write functions that work on any object that has certain methods. The function doesn’t care about the exact class of the object – only that it can perform the required actions.

```python
def print_marksheet(student):
    """This function works with any student object that has a display_info method."""
    print("\n--- Printing Mark Sheet ---")
    student.display_info()
    print("--------------------------\n")

# Call with different types of students
print_marksheet(science)
print_marksheet(commerce)
print_marksheet(humanities)
```


---

## 4. Duck Typing – “If it walks like a duck and quacks like a duck…”  

Python is dynamically typed, so it doesn’t check the type of an object before calling a method. If an object has the method, Python just calls it. This is called **duck typing**.

Let’s create a completely different class that also has a `display_info` method, but is not related to `Student` at all.

```python
class Teacher:
    def __init__(self, name, subject, experience):
        self.name = name
        self.subject = subject
        self.experience = experience

    def display_info(self):
        print(f"Teacher: {self.name}")
        print(f"Subject: {self.subject}")
        print(f"Experience: {self.experience} years")
```

```python
teacher = Teacher("Mr. Radhakrishnan", "Mathematics", 15)
```

Now, our `print_marksheet` function works with the teacher too – because it only requires a `display_info` method.

```python
print_marksheet(teacher)   # No error, even though Teacher is not a Student!
```

> That’s duck typing: Python doesn’t care about the object’s class; if it has the method, it’s acceptable.

## 5. Built‑in Example of Polymorphism  

Many built‑in functions in Python are polymorphic. For instance, `len()` works on strings, lists, tuples, dictionaries – all different types.

```python
print(len("hello"))        # string
print(len([1, 2, 3, 4]))   # list
print(len((10, 20)))       # tuple
print(len({"a":1, "b":2})) # dictionary
```

Each type implements `__len__` in its own way, but we can call `len()` on all of them.

## 6. Polymorphism with the Same Method Name Across Unrelated Classes  

Let’s create another class that has a `total_marks` method, just for fun.

```python
class Exam:
    def __init__(self, subject, max_marks, obtained):
        self.subject = subject
        self.max_marks = max_marks
        self.obtained = obtained

    def total_marks(self):
        return self.obtained
```

```python
def show_total(obj):
    print(f"Total marks: {obj.total_marks()}")

show_total(science)      # Student object
show_total(Exam("Maths", 100, 95))   # Exam object
```


Both objects work because they provide a `total_marks` method. The function doesn’t need to know their actual types.

---

## 7. Why Polymorphism Makes Code Flexible  

- You can write code that works on a **family of classes** (like all student streams) without knowing each one’s details.  
- New classes can be added later without changing existing functions – just ensure they implement the required methods.  
- It encourages you to think in terms of **what an object can do** rather than **what it is**.

---

## Summary  

- **Polymorphism** means “many forms” – the same method name can have different implementations in different classes.  
- **Method overriding** (in inheritance) is one way to achieve polymorphism.  
- **Duck typing** is another: Python doesn’t check the type of an object; it just tries to call the method. If the object has it, it works.  
- Polymorphism allows you to write more **flexible, reusable** code.  
- Many built‑in functions (`len`, `max`, etc.) are polymorphic.
