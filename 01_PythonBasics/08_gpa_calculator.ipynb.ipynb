# GPA Calculator – Kerala SSLC Style

Welcome! In this notebook, we'll build a simple GPA calculator using the Kerala SSLC grade system.

You'll practice:
- **Lists** to store subjects and marks
- **Loops** to process data
- **Conditionals** to map marks to grades
- **Dictionaries** for grade mappings
- **Functions** to organize code

By the end, you'll calculate your own GPA from your mark sheet! 😊

---

## Step 1: Store Subject Names and Marks

We'll start by creating two lists:
1. **Subject names** – the courses you took
2. **Marks** – the score you got in each (out of 100)

Let's use a real example with 6 SSLC subjects.

```python
# List of subject names
subjects = ["Malayalam", "English", "Hindi", "Mathematics", "Science", "Social Science"]

# Your marks (out of 100) – replace with your own!
marks = [95, 88, 72, 98, 85, 78]

# Print to see the data
print("Subjects:", subjects)
print("Marks   :", marks)
```

---

## Step 2: Calculate Total and Percentage

Now let's compute:
- **Total marks** – sum of all marks  
- **Percentage** – average score (since each subject is out of 100)

We'll use `sum()` and `len()` functions.

```python
# Calculate total marks
total_marks = sum(marks)

# Number of subjects
num_subjects = len(subjects)

# Average percentage (each subject is out of 100)
percentage = total_marks / num_subjects

print(f"Total marks: {total_marks} / {num_subjects * 100}")
print(f"Percentage : {percentage:.2f}%")
```

---

## Step 3: Map Marks to Grade Points

In Kerala SSLC, marks are converted to **grade points** using this scale:

| Marks Range | Grade | Grade Point |
|:---:|:---:|:---:|
| 90–100 | A+ | 9 |
| 80–89 | A | 8 |
| 70–79 | B+ | 7 |
| 60–69 | B | 6 |
| 50–59 | C+ | 5 |
| 40–49 | C | 4 |
| 30–39 | D+ | 3 |
| 20–29 | D | 2 |
| 0–19 | E | 1 |

We'll loop through each mark and assign the corresponding grade point using `if`/`elif` statements.

```python
# Create empty list to store grade points
grade_points = []

# Loop through each mark and assign grade point
for mark in marks:
    if mark >= 90:
        gp = 9
    elif mark >= 80:
        gp = 8
    elif mark >= 70:
        gp = 7
    elif mark >= 60:
        gp = 6
    elif mark >= 50:
        gp = 5
    elif mark >= 40:
        gp = 4
    elif mark >= 30:
        gp = 3
    elif mark >= 20:
        gp = 2
    else:
        gp = 1
    
    grade_points.append(gp)

# Display subject-wise grade points
print("Subject-wise Grade Points:")
for i in range(num_subjects):
    print(f"  {subjects[i]:15} : {marks[i]:3} marks → Grade point {grade_points[i]}")
```

---

## Step 4: Calculate GPA

**GPA** (Grade Point Average) is simply the average of all grade points.

Formula: `GPA = Sum of all grade points / Number of subjects`

```python
# Calculate GPA
gpa = sum(grade_points) / num_subjects

print("\n=== GPA Summary ===")
print(f"Number of subjects : {num_subjects}")
print(f"Grade points       : {grade_points}")
print(f"GPA                : {gpa:.2f}")
```

---

## Step 5: Create a Reusable Function

Let's wrap all the logic into a function that we can use anytime.

This function takes:
- **Input**: subject names and marks
- **Output**: calculated GPA

```python
def calculate_gpa(subjects, marks):
    """
    Calculate GPA from subject marks using Kerala SSLC grade scale.
    
    Args:
        subjects (list): List of subject names
        marks (list): List of marks (0-100)
    
    Returns:
        float: Calculated GPA
    """
    # Convert marks to grade points
    grade_points = []
    
    for m in marks:
        if m >= 90:
            gp = 9
        elif m >= 80:
            gp = 8
        elif m >= 70:
            gp = 7
        elif m >= 60:
            gp = 6
        elif m >= 50:
            gp = 5
        elif m >= 40:
            gp = 4
        elif m >= 30:
            gp = 3
        elif m >= 20:
            gp = 2
        else:
            gp = 1
        
        grade_points.append(gp)
    
    # Calculate and return GPA
    gpa = sum(grade_points) / len(marks)
    return gpa

# Test the function with our data
my_gpa = calculate_gpa(subjects, marks)
print(f"Your GPA is: {my_gpa:.2f}")
```

---

## Step 6: Use Grades Instead of Marks

Sometimes you already have letter grades (A+, A, B+, etc.) instead of marks.

We can use a **dictionary** to map grades directly to grade points, then calculate GPA.

```python
# Dictionary mapping grades to grade points
grade_to_point = {
    "A+": 9,
    "A": 8,
    "B+": 7,
    "B": 6,
    "C+": 5,
    "C": 4,
    "D+": 3,
    "D": 2,
    "E": 1
}

# Example: grades for the same subjects
grades = ["A+", "A", "B+", "A+", "A", "B+"]

# Convert grades to grade points using dictionary lookup
# (This is a list comprehension – a compact way to build a list)
grade_points_from_grades = [grade_to_point[g] for g in grades]

# Calculate GPA
gpa_from_grades = sum(grade_points_from_grades) / len(grades)

print("Grades entered    :", grades)
print("Corresponding points:", grade_points_from_grades)
print(f"GPA from grades   : {gpa_from_grades:.2f}")
```

**Note:** The list comprehension `[grade_to_point[g] for g in grades]` is a compact way to build a list. It's like a short `for` loop. If you're not comfortable with it yet, you can write a regular loop instead.

---

## Step 7: Handle Any Number of Subjects

Our code works with any number of subjects because we use `len()` and loops.

Try testing with a different number of subjects. Our function adapts automatically!

```python
# Test with only 3 subjects
short_subjects = ["Mathematics", "Physics", "Chemistry"]
short_marks = [100, 89, 76]

gpa_short = calculate_gpa(short_subjects, short_marks)

print(f"Test with {len(short_subjects)} subjects:")
for i, (subj, mark) in enumerate(zip(short_subjects, short_marks)):
    print(f"  {i+1}. {subj}: {mark}")

print(f"\nGPA for 3 subjects: {gpa_short:.2f}")
```
