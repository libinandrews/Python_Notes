# BMI Calculator with Personalized Advice

Welcome! In this notebook, we'll build a **Body Mass Index (BMI) calculator** with health recommendations.

## Step 1: Store Your Height and Weight

First, we'll store your weight and height measurements.
You can edit these values with your own data.

**Note:** We'll use metric units (kg and metres) in this step.
- Weight in **kilograms** (kg)
- Height in **metres** (m)

```python
# 👇 EDIT THESE VALUES WITH YOUR OWN MEASUREMENTS
weight_kg = 70        # weight in kilograms
height_m = 1.75       # height in metres

# Display the values
print("Your measurements:")
print(f"  Weight: {weight_kg} kg")
print(f"  Height: {height_m} m")
```

---

## Step 2: Calculate BMI

**BMI** (Body Mass Index) is a measure of body fat based on height and weight.

**Formula:**
$$\text{BMI} = \frac{\text{weight (kg)}}{\text{height (m)}^2}$$

For example, with weight = 70 kg and height = 1.75 m:
$$\text{BMI} = \frac{70}{1.75^2} = \frac{70}{3.0625} \approx 22.86$$

```python
# Calculate BMI using the formula
bmi = weight_kg / (height_m ** 2)

# Display the result (rounded to 2 decimal places)
print(f"Your BMI is: {bmi:.2f}")
```

---

## Step 3: Classify BMI Category

The World Health Organization (WHO) defines these BMI categories for adults:

| BMI Range | Category |
|:---:|:---|
| Less than 18.5 | Underweight |
| 18.5 – 24.9 | Normal |
| 25.0 – 29.9 | Overweight |
| 30.0 or more | Obese |

We'll use `if`/`elif`/`else` statements to determine which category your BMI falls into.

```python
# Determine BMI category using if/elif/else
if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:           # 18.5 ≤ bmi < 25
    category = "Normal"
elif bmi < 30:           # 25 ≤ bmi < 30
    category = "Overweight"
else:                    # bmi ≥ 30
    category = "Obese"

print(f"Your BMI category is: {category}")
```

---

## Step 4: Create a Function with Personalized Advice

Now let's wrap the logic into a reusable **function** that returns:
1. **BMI** – the calculated value
2. **Category** – the classification
3. **Advice** – personalized health recommendations

```python
def bmi_advice(weight, height):
    """
    Calculate BMI and return personalized health advice.
    
    Args:
        weight (float): Weight in kilograms
        height (float): Height in metres
    
    Returns:
        tuple: (bmi, category, advice)
    """
    # Calculate BMI
    bmi = weight / (height ** 2)
    
    # Determine category and advice
    if bmi < 18.5:
        category = "Underweight"
        advice = "You may need to gain some weight. Consider a balanced diet with nutrient-rich foods."
    elif bmi < 25:
        category = "Normal"
        advice = "Great! You are in a healthy weight range. Keep up the good habits."
    elif bmi < 30:
        category = "Overweight"
        advice = "You might benefit from a bit more physical activity and a balanced diet."
    else:
        category = "Obese"
        advice = "Consider consulting a healthcare provider for a personalized wellness plan."
    
    return bmi, category, advice

# Test the function with our data
my_bmi, my_category, my_advice = bmi_advice(weight_kg, height_m)

print("=== Your BMI Results ===")
print(f"BMI: {my_bmi:.2f}")
print(f"Category: {my_category}")
print(f"Advice: {my_advice}")
```

## Step 5: Support Imperial Units (Pounds & Inches)

In some countries, people use imperial units:
- Weight in **pounds** (lb)
- Height in **inches** (or feet and inches)

To support both systems, we can modify our function to accept a `unit` parameter.

**Conversion factors:**
- 1 pound = 0.453592 kilograms
- 1 inch = 0.0254 metres

```python
def bmi_advice_units(weight, height, unit='metric'):
    """
    Calculate BMI with unit selection.
    
    Args:
        weight (float): Weight value
        height (float): Height value
        unit (str): 'metric' (kg, m) or 'imperial' (lb, inches)
    
    Returns:
        tuple: (bmi, category, advice)
    """
    # Convert to metric if needed
    if unit == 'imperial':
        weight_kg = weight * 0.453592      # pounds to kg
        height_m = height * 0.0254         # inches to metres
    else:
        weight_kg = weight
        height_m = height
    
    # Calculate BMI
    bmi = weight_kg / (height_m ** 2)
    
    # Determine category and advice (same logic as before)
    if bmi < 18.5:
        category = "Underweight"
        advice = "You may need to gain some weight. Consider a balanced diet with nutrient-rich foods."
    elif bmi < 25:
        category = "Normal"
        advice = "Great! You are in a healthy weight range. Keep up the good habits."
    elif bmi < 30:
        category = "Overweight"
        advice = "You might benefit from a bit more physical activity and a balanced diet."
    else:
        category = "Obese"
        advice = "Consider consulting a healthcare provider for a personalized wellness plan."
    
    return bmi, category, advice

# Example: imperial units (154 lb, 68 inches = 5'8")
weight_lb = 154
height_inches = 68

bmi_imp, cat_imp, adv_imp = bmi_advice_units(weight_lb, height_inches, unit='imperial')

print("=== Imperial Units Example ===")
print(f"Weight: {weight_lb} lb, Height: {height_inches} inches")
print(f"BMI: {bmi_imp:.2f}")
print(f"Category: {cat_imp}")
print(f"Advice: {adv_imp}")
```
