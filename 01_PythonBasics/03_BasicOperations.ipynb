# String Operations
### 1. **String Concatenation**
Combining two or more strings.


```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)  # Output: John Doe
```

### 2. **String Interpolation (f-strings)**
Embedding expressions inside string literals.

```python
name = "Alice"
age = 25
greeting = f"My name is {name} and I am {age} years old."
print(greeting)  # Output: My name is Alice and I am 25 years old.
```

### 3. **String Formatting**
Using the `.format()` method.

```python
name = "Bob"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)  # Output: My name is Bob and I am 30 years old.
```

### 4. **String Slicing**
Extracting a portion of a string.

```python
text = "Hello, World!"
print(text[0:5])  # Output: Hello
print(text[7:])   # Output: World!
print(text[-6:-1])  # Output: World
```


### 5. **String Length**
Getting the length of a string.

```python
text = "Python"
print(len(text)) 
```

### 6. **String Case Conversion**
Changing the case of a string.

```python
text = "Hello, World!"
print(text.upper())  # Output: HELLO, WORLD!
print(text.lower())  # Output: hello, world!
print(text.title())  # Output: Hello, World!
```

### 7. **String Stripping**
Removing leading and trailing whitespace.

```python
text = "   Python   "
print(text.strip())  # Output: Python
print(text.lstrip())  # Output: Python   (left strip)
print(text.rstrip())  # Output:    Python (right strip)
```

### 8. **String Splitting**
Splitting a string into a list.

```python
text = "apple,banana,cherry"
fruits = text.split(",")
print(fruits)
```

### 9. **String Joining**
Combining a list of strings into a single string.

```python
fruits = ["apple", "banana", "cherry"]
text = ", ".join(fruits)
print(text) 
```

### 10. **String Replacement**
Replacing substrings.

```python
text = "Hello, World!"
new_text = text.replace("World", "Python")
print(new_text) 
```

### 11. **String Searching**
Checking if a substring exists.

```python
text = "Hello, World!"
print("World" in text)
print("Python" in text)
```

### 12. **Finding Substring Index**
Finding the index of a substring.


```python
text = "Hello, World!"
print(text.find("World"))  # Output: 7
print(text.find("Python"))  # Output: -1 (not found)
```

### 13. **Multiline Strings**
Using triple quotes for multiline strings.

```python
text = """Hello,
World!"""
print(text)
```

### 14. **String Checking**
Checking if a string starts or ends with a substring.

```python
text = "Hello, World!"
print(text.startswith("Hello"))  # Output: True
print(text.endswith("World!"))  
```

### 15. **String Checking Methods**
Checking if a string meets certain conditions.


```python
text = "Python123"
print(text.isalnum())  # Output: True (alphanumeric)
print(text.isalpha())  # Output: False (alphabetic only)
print(text.isdigit())  # Output: False (digits only)
print(text.islower())  # Output: False (all lowercase)
print(text.isupper())  # Output: False (all uppercase)
```

# **Assignment Operations**

 - **Assign (`=`)**: Assigns a value to a variable.

```python
 x = 10
 print(x) 
```

- **Add and Assign (`+=`)**: Adds and assigns the result.

```python
x += 5
print(x) 
```

- **Subtract and Assign (`-=`)**: Subtracts and assigns the result.

```python
x -= 3
print(x) 
```

- **Multiply and Assign (`*=`)**: Multiplies and assigns the result.

```python
x *= 2
print(x) 
```

  - **Divide and Assign (`/=`)**: Divides and assigns the result.

```python
x /= 4
print(x) 
```

   - **Floor Divide and Assign (`//=`)**: Floor divides and assigns the result.

```python
x //= 2
print(x)
```

   - **Modulus and Assign (`%=`)**: Takes modulus and assigns the result.

```python
x %= 2
print(x) 
```

   - **Exponentiate and Assign (`**=`)**: Raises to the power and assigns the result.

```python
x **= 3
print(x) 
```
