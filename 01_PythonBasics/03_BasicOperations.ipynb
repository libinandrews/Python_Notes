# Comparison & Logical Operators
Now that we know about basic data types, let's see how we can **compare values** and make **decisions** in Python. This is where programming gets really useful!

Think about your mark sheet - how did the school decide if you passed or failed? They had to check each subject's grade against the passing criteria. That's exactly what comparison and logical operators help us do.


Comparison Operators

Comparison operators compare two values and give us a `Boolean` result - either `True` or `False`.

### The Comparison Operators

| Operator | Meaning | Example | Result (based on your marks) |
|:---:|:---|:---|:---|
| `==` | Equal to | `"LAALU" == "LAALU"` | `True` |
| `!=` | Not equal to | `"LAALU" != "RAHEEM"` | `True` |
| `>` | Greater than | `85 > 75` | `True` |
| `<` | Less than | `85 < 75` | `False` |
| `>=` | Greater than or equal to | `85 >= 85` | `True` |
| `<=` | Less than or equal to | `85 <= 90` | `True` |


````python
# Let's store some grades from your mark sheet
malayalam_grade = "B+" 
english_grade = "B+"
math_grade = "B+"
physics_grade = "B+"
chemistry_grade = "A"
social_grade = "A"

# Let's also store the actual percentages (assuming these are your scores)
malayalam_score = 75.5   # B+ range (70-79%)
chemistry_score = 85.0   # A range (80-89%)
math_score = 71.0        # B+ range
passing_score = 40       # Minimum C grade is 40%
````

````python
# Example 1: Check if you passed Malayalam
passed_malayalam = malayalam_score >= passing_score
print(f"Passed Malayalam? {passed_malayalam}")  # True
````

````python
# Example 2: Check if you got an A in Chemistry
got_A_in_chemistry = chemistry_score >= 80
print(f"Got A in Chemistry? {got_A_in_chemistry}")  # True
````

````python
# Example 3: Compare two scores
is_chemistry_higher_than_math = chemistry_score > math_score
print(f"Is Chemistry score higher than Math? {is_chemistry_higher_than_math}")  # True
````

````python
# Example 4: Check exact grade
is_math_grade_B_plus = (math_grade == "B+")
print(f"Is Math grade B+? {is_math_grade_B_plus}")  # True
````

````python
# Example 5: Not equal example
is_physics_not_A = (physics_grade != "A")
print(f"Is Physics NOT an A? {is_physics_not_A}")  # True (since it's B+)
````

````python
# Example 6: Check if any score is below passing
failed_any = malayalam_score < passing_score
print(f"Failed any subject? {failed_any}")  # False
````

## Logical Operators

Logical operators let us combine multiple conditions. They're like asking "and...", "or...", or "not..." questions in English.

### The Logical Operators

| Operator | Meaning | Example | Explanation |
|:---:|:---|:---|:---|
| `and` | Both conditions must be True | `(score >= 40) and (score < 50)` | Is score in C grade range? |
| `or` | At least one condition must be True | `(grade == "A+") or (grade == "A")` | Is this an A-grade? |
| `not` | Reverses the condition | `not passed` | Did they fail? |

### Truth Tables

#### AND (`and`)
| Condition 1 | Condition 2 | Result |
|:---:|:---:|:---:|
| `True` | `True` | `True` |
| `True` | `False` | `False` |
| `False` | `True` | `False` |
| `False` | `False` | `False` |
---
#### OR (`or`)
| Condition 1 | Condition 2 | Result |
|:---:|:---:|:---:|
| `True` | `True` | `True` |
| `True` | `False` | `True` |
| `False` | `True` | `True` |
| `False` | `False` | `False` |
---
#### NOT (`not`)
| Condition | Result |
|:---:|:---:|
| `True` | `False` |
| `False` | `True` |


````python
# Your marks data
malayalam = 75
english = 74
math = 71
chemistry = 82
pass_mark = 40
distinction_mark = 80
````

````python
# AND examples - both conditions must be True
print(malayalam >= pass_mark and english >= pass_mark)  # True - both passed
print(math >= 80 and chemistry >= 80)                   # False - math < 80
print(malayalam > 70 and english > 70)                  # True - both > 70
````

````python
# OR examples - at least one condition must be True
print(malayalam >= 80 or chemistry >= 80)               # True - chemistry >= 80
print(math >= 80 or english >= 80)                       # False - neither >= 80
print(malayalam < 40 or english < 40)                    # False - both passed
````

````python
# NOT examples - reverses the condition
passed_all = True
failed_any = False

print(not passed_all)        # False - reverses True to False
print(not failed_any)        # True - reverses False to True
print(not (math < 40))       # True - math < 40 is False, so not False = True
````

# String Operations
### 1. **String Concatenation**
Combining two or more strings.


````python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)  # Output: John Doe
````

### 2. **String Interpolation (f-strings)**
Embedding expressions inside string literals.

````python
name = "Alice"
age = 25
greeting = f"My name is {name} and I am {age} years old."
print(greeting)  # Output: My name is Alice and I am 25 years old.
````

### 3. **String Formatting**
Using the `.format()` method.

````python
name = "Bob"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)  # Output: My name is Bob and I am 30 years old.
````

### 4. **String Slicing**
Extracting a portion of a string.

````python
text = "Hello, World!"
print(text[0:5])  # Output: Hello
print(text[7:])   # Output: World!
print(text[-6:-1])  # Output: World
````


### 5. **String Length**
Getting the length of a string.

````python
text = "Python"
print(len(text)) 
````

### 6. **String Case Conversion**
Changing the case of a string.

````python
text = "Hello, World!"
print(text.upper())  # Output: HELLO, WORLD!
print(text.lower())  # Output: hello, world!
print(text.title())  # Output: Hello, World!
````

### 7. **String Stripping**
Removing leading and trailing whitespace.

````python
text = "   Python   "
print(text.strip())  # Output: Python
print(text.lstrip())  # Output: Python   (left strip)
print(text.rstrip())  # Output:    Python (right strip)
````

### 8. **String Splitting**
Splitting a string into a list.

````python
text = "apple,banana,cherry"
fruits = text.split(",")
print(fruits)
````

### 9. **String Joining**
Combining a list of strings into a single string.

````python
fruits = ["apple", "banana", "cherry"]
text = ", ".join(fruits)
print(text) 
````

### 10. **String Replacement**
Replacing substrings.

````python
text = "Hello, World!"
new_text = text.replace("World", "Python")
print(new_text) 
````

### 11. **String Searching**
Checking if a substring exists.

````python
text = "Hello, World!"
print("World" in text)
print("Python" in text)
````

### 12. **Finding Substring Index**
Finding the index of a substring.


````python
text = "Hello, World!"
print(text.find("World"))  # Output: 7
print(text.find("Python"))  # Output: -1 (not found)
````

### 13. **Multiline Strings**
Using triple quotes for multiline strings.

````python
text = """Hello,
World!"""
print(text)
````

### 14. **String Checking**
Checking if a string starts or ends with a substring.

````python
text = "Hello, World!"
print(text.startswith("Hello"))  # Output: True
print(text.endswith("World!"))  
````

### 15. **String Checking Methods**
Checking if a string meets certain conditions.


````python
text = "Python123"
print(text.isalnum())  # Output: True (alphanumeric)
print(text.isalpha())  # Output: False (alphabetic only)
print(text.isdigit())  # Output: False (digits only)
print(text.islower())  # Output: False (all lowercase)
print(text.isupper())  # Output: False (all uppercase)
````
