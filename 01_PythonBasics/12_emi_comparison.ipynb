# EMI vs. Full Payment vs. No-Cost EMI

Welcome! After your SSLC exams, you might be thinking about buying a new laptop for your higher secondary studies. Suppose you find a laptop priced at ₹58,570. You have three payment options:

- **Pay the full amount upfront** – simple, no extra cost.

- **Regular EMI** – you pay in monthly installments with interest on the reducing balance.

- **No-cost EMI** – you pay the price in equal monthly installments without interest, but there might be a hidden processing fee.

In this notebook, we'll build a Python program to compare these three options and understand the true cost of credit. We'll use only the Python concepts you already know: data types, operations, lists, dictionaries, loops, functions, and a little bit of `input()`.

Let's dive in!
