# Day 04 **Data Structures**
**Data structures** are specialized formats for organizing, processing, and storing data. Python has several built-in data structures that make it easy to work with collections of information – like the items on your Kerala SSLC mark sheet!

## 1️⃣ Lists (`list`)

A **list** is an ordered, mutable (changeable) collection of items. Items can be of any data type, and you can add, remove, or modify them.

```python
# Creating a List
subjects = ["Malayalam", "English", "Hindi", "Social Science", "Physics", "Chemistry", "Biology", "Mathematics", "Information Technology"]
print(subjects)
```

```python
# Accessing Items (Indexing)
# Index starts at 0
first_subject = subjects[0]       # "Malayalam"
third_subject = subjects[2]        # "Hindi"
last_subject = subjects[-1]        # "Information Technology" (negative index counts from end)

print("First:", first_subject)
print("Third:", third_subject)
print("Last:", last_subject)
```

```python
# Slicing – Get a sublist
# First three subjects
first_three = subjects[0:3]        # from index 0 up to (but not including) 3
print("First three:", first_three)

# Last two subjects
last_two = subjects[-2:]           # from second-last to end
print("Last two:", last_two)
```

```python
# Modifying a List
# You had 9 subjects, but suppose you want to rename "Information Technology" to "IT"
subjects[-1] = "IT"
print("Updated subjects:", subjects)
```

```python
# Adding Items
# Add a new subject (optional)
subjects.append("Physical Education")
print("After append:", subjects)
```

```python
# Removing Items
# Remove by value
subjects.remove("Physical Education")
print("After remove:", subjects)
```

```python
# Remove by index (pop)
last = subjects.pop()      # removes and returns last item
print("Popped:", last)
print("Now subjects:", subjects)
```

```python
# List Operations
# Length of list
num_subjects = len(subjects)
print("Number of subjects:", num_subjects)
```

## 2️⃣ Tuples (`tuple`)
A **tuple** is an ordered, **immutable** (cannot be changed) collection. Once created, you cannot add, remove, or modify items. Use tuples for data that should never change.
### When to Use Tuple
- Fixed data (like DOB, registration number)
- Dictionary keys (only immutable types can be keys)
- Better performance than list for read-only data

```python
# Creating a Tuple
# Your date of birth (fixed, never changes)
dob = (22, "July", 1985)
print("Date of Birth:", dob)

# Your exam details (register number, month, year)
exam_details = ("006615", "March", 2021)
print("Exam details:", exam_details)
```

```python
# Accessing Tuple Items
day = dob[0]
month = dob[1]
year = dob[2]

print(f"Born on {day} {month} {year}")
```

```python
# Trying to Modify (Will Cause ❌ TypeError)
# dob[0] = 23 
```

## 3️⃣ Dictionaries (`dict`)


A **dictionary** stores data in **key-value pairs**. It's unordered (in older Python) but ordered from Python 3.7+. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
# Creating a Dictionary
# Subject names as keys, grades as values
grades = {
    "Malayalam": "B+",
    "English": "B+",
    "Hindi": "B+",
    "Social Science": "A",
    "Physics": "B+",
    "Chemistry": "A",
    "Biology": "B",
    "Mathematics": "B+",
    "Information Technology": "A"
}
print("Grades dictionary:")
print(grades)
```

```python
# Accessing Values
# Using key
math_grade = grades["Mathematics"]
print("Math grade:", math_grade)

# Using get() (safe, returns None if key missing)
chem_grade = grades.get("Chemistry")
print("Chemistry grade:", chem_grade)
```

```python
# Adding/Updating Key-Value Pairs
# Add a new subject (if you had an additional subject)
grades["Sanskrit"] = "A"
print("After adding Sanskrit:", grades)

# Update an existing grade
grades["Mathematics"] = "A"
print("After updating Math:", grades["Mathematics"])
```

```python
# Removing Key-Value Pairs
# Remove by key
removed = grades.pop("Sanskrit")
print("Removed:", removed)
print("After pop:", grades)
```

```python
# Dictionary Operations
# Number of entries
print("Total subjects:", len(grades))

# Check if a key exists
if "Physics" in grades:
    print("Physics grade found.")
```


## 4️⃣ Sets (`set`)

A **set** is an unordered collection of **unique** items. It's mutable, but each element must be immutable. Useful for removing duplicates and mathematical set operations.
#### When to Use Set
- Remove duplicates from a list
- Test membership quickly (faster than list)
- Mathematical operations like union, intersection

```python
# Creating a Set
# All grades you received (without duplicates)
unique_grades = {"B+", "A", "B"}   # Notice no duplicates
print("Unique grades:", unique_grades)
```

```python
# Adding and Removing
# Add a grade (if you had an A+)
unique_grades.add("A+")
print("After add:", unique_grades)

# Remove a grade
unique_grades.remove("B")
print("After remove:", unique_grades)
```

```python
# Set Operations
# Subjects you got A in
a_subjects = {"Social Science", "Chemistry", "Information Technology"}
# Subjects you got B+ in
bplus_subjects = {"Malayalam", "English", "Hindi", "Physics", "Mathematics"}
```

```python
# Union – all subjects with A or B+
all_good = a_subjects.union(bplus_subjects)
print("All A or B+ subjects:", all_good)
```

```python
# Intersection – subjects that are both? (none here)
common = a_subjects.intersection(bplus_subjects)
print("Common subjects:", common)   # empty set
```

```python
# Difference – subjects with A but not B+
only_a = a_subjects.difference(bplus_subjects)
print("Only A subjects:", only_a)
```

## 📊 Summary Table: Python Data Structures

| Structure | Ordered | Mutable | Duplicates | When to Use | Mark Sheet Example |
|:---|:---:|:---:|:---:|:---|:---|
| **List** | ✅ | ✅ | ✅ | General collection, need to modify | `["Malayalam", "English", ...]` |
| **Tuple** | ✅ | ❌ | ✅ | Fixed data, protect from changes | `(22, "July", 1985)` |
| **Dictionary** | ✅* | ✅ | Keys: ❌, Values: ✅ | Key-value lookups | `{"Malayalam": "B+", ...}` |
| **Set** | ❌ | ✅ | ❌ | Unique items, set operations | `{"A", "B+", "B"}` |
| **String** | ✅ | ❌ | ✅ | Text data | `"LAALU"` |


\* Dictionaries maintain insertion order from Python 3.7+

## 🔧 Quick Reference – Common Operations

| Operation | List | Tuple | Dict | Set | String |
|:---|:---:|:---:|:---:|:---:|:---:|
| Create | `[1,2]` | `(1,2)` | `{"a":1}` | `{1,2}` | `"abc"` |
| Access | `l[0]` | `t[0]` | `d["a"]` | - (no index) | `s[0]` |
| Add | `l.append(3)` | ❌ | `d["b"]=2` | `s.add(3)` | ❌ |
| Remove | `l.remove(2)` | ❌ | `d.pop("a")` | `s.remove(2)` | ❌ |
| Length | `len(l)` | `len(t)` | `len(d)` | `len(s)` | `len(s)` |
| Loop | `for i in l:` | `for i in t:` | `for k in d:` | `for i in s:` | `for ch in s:` |
| Membership | `2 in l` | `2 in t` | `"a" in d` | `2 in s` | `"b" in s` |

## 🧪 Putting It All Together – Complete Mark Sheet Data Structure

```python
mark_sheet = {
    "register_number": "006615",
    "candidate_name": "LAALU",
    "father_name": '"WELL-DONE" VASU',
    "date_of_birth": (22, "July", 1985),                 # tuple
    "school": "DESAM GOVERNMENT HIGH SCHOOL",
    "exam_centre": "DESAM PUBLIC LIBRARY (Temporary)",
    "month_year": ("March", 2021),                       # tuple
    "chances": 6,
    "result": "PASSED",
    "subjects": ["Malayalam", "English", "Hindi", "Social Science",    # list
                 "Physics", "Chemistry", "Biology", "Mathematics", 
                 "Information Technology"],
    "grades": {                                           # dictionary
        "Malayalam": "B+",
        "English": "B+",
        "Hindi": "B+",
        "Social Science": "A",
        "Physics": "B+",
        "Chemistry": "A",
        "Biology": "B",
        "Mathematics": "B+",
        "Information Technology": "A"
    },
    "unique_grades": {"A", "B+", "B"}                     # set
}

# Accessing data
print("Candidate:", mark_sheet["candidate_name"])
print("Born on:", mark_sheet["date_of_birth"][0], mark_sheet["date_of_birth"][1])
print("Math grade:", mark_sheet["grades"]["Mathematics"])
print("All unique grades:", mark_sheet["unique_grades"])
```

## ✅ Key Takeaways

- **Lists** are flexible and most commonly used.
- **Tuples** protect data that shouldn't change.
- **Dictionaries** are perfect for paired data (like subject → grade).
- **Sets** help with unique values and mathematical operations.
- **Strings** are sequences too, with their own methods.
