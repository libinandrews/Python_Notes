## 1. Why Organize Code?

As your programs grow, putting everything in one file becomes messy. Python gives us tools to split code into separate files and folders:

- **Function**: A reusable block of code (you already know this).
- **Module**: A single `.py` file that contains functions, variables, etc.
- **Package**: A folder that contains multiple modules (and a special `__init__.py` file).

Think of it like your school notes:
- A function is like a single formula.
- A module is like a chapter covering one topic.
- A package is like a whole subject folder with multiple chapters.

To use code from a module or package, we use the `import` statement. Python provides several ways to import:

```python
import module_name                 # access with module_name.function()
from module_name import function   # use function directly
import module_name as mn           # give it a shorter alias
```

Now let's see a quick example before diving into our project.

`````python
# Import the whole module
import math_helpers
`````

`````python
print(math_helpers.square(5))
print(math_helpers.cube(3))
`````

`````python
# Import specific functions
from math_helpers import square, cube

print(square(4))
print(cube(2))
`````

`````python
# Use an alias
import math_helpers as mh

print(mh.square(6))
print(mh.cube(2))
`````

## Relative Imports

- Relative imports are used to import modules **within the same package**.
- They help organize large projects where modules are stored inside folders.
- Instead of writing the full package path, we use **dots (`.`)** to indicate location.

### Meaning of dots
- `.` → current package
- `..` → parent package
- `...` → grandparent package

### Example folder structure

## Installing Python Libraries

### What is a Python Library?

- A **Python library** is a collection of pre-written code that helps perform specific tasks.
- Libraries save time by providing ready-made functions and tools.
- Examples of popular libraries:
  - `numpy` → numerical computing
  - `pandas` → data analysis
  - `matplotlib` → data visualization

---

## Installing Libraries using `pip`

- `pip` is the **default package manager for Python**.
- It allows you to install libraries from the **Python Package Index (PyPI)**.

### Basic Syntax

```bash
pip install library_name
````

Example:

```bash
pip install numpy
```

---

## Installing a Specific Version

Sometimes a project requires a specific version of a library.

```bash
pip install library_name==version
```

Example:

```bash
pip install numpy==1.26.0
```

---

## Installing Multiple Libraries

You can install multiple libraries at once.

```bash
pip install numpy pandas matplotlib
```

---

## Checking Installed Libraries

To see all installed packages:

```bash
pip list
```

---

## Installing from `requirements.txt`

Projects often store dependencies in a file called `requirements.txt`.

Example file:

```
numpy
pandas
matplotlib
```

Install all libraries using:

```bash
pip install -r requirements.txt
```

### Python Code Organization Summary

| Level | Physical Representation | Analogy | Best Used For... |
| --- | --- | --- | --- |
| **Function** | A block of code (`def`). | A single **instruction** or specific step. | Repeating a specific task within your script. |
| **Module** | A single `.py` file. | A single **tool** (e.g., a hammer). | Organizing related functions and classes in one file. |
| **Package** | A folder with `__init__.py`. | A **toolbox** containing tools. | Grouping related modules into a hierarchy. |
| **Library** | A collection of packages/modules. | A **hardware store** of toolboxes. | Providing a full suite of features for a specific field. |

---

### Key Hierarchy

Think of it as a nesting doll:

1. **Functions** live inside **Modules**.
2. **Modules** live inside **Packages**.
3. **Packages** (and modules) are collected into **Libraries**.
