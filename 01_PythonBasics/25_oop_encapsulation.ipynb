# 25. Object-Oriented Programming: Encapsulation  
## Protecting Your Kerala SSLC Mark Sheet  

Welcome! In this notebook, we’ll learn about **encapsulation** – one of the core ideas of object‑oriented programming.  
Encapsulation helps us protect an object’s internal data from accidental or unwanted changes.  

We’ll continue with our SSLC mark sheet example. Imagine you have a student’s marks. You don’t want anyone to change those marks directly without proper checks – after all, marks are sensitive! Encapsulation gives us a way to control how data is accessed and modified.

---

## 1. What is Encapsulation?  

- **Encapsulation** means bundling data (attributes) and methods (behaviour) together inside a class, while **restricting direct access** to some of the object’s internal parts.  
- It’s like a **protective wrapper** around the data.  
- In Python, we achieve encapsulation by using:  
  - **Public attributes** – accessible from anywhere.  
  - **Protected attributes** – by convention, meant for internal use (single underscore `_`).  
  - **Private attributes** – name mangling makes them harder to access directly (double underscore `__`).  

> 🔒 **Why encapsulate?**  
> - Prevents accidental modification of important data.  
> - Allows you to add validation or logic when data is changed.  
> - Makes your code more reliable and easier to maintain.

---

## 2. Public Attributes – The Default  

By default, all attributes in a Python class are **public**. That means they can be read and changed directly from outside the class.


```python
class Student:
    def __init__(self, name, marks):
        self.name = name      # public attribute
        self.marks = marks    # public attribute

    def total_marks(self):
        return sum(self.marks)

# Create a student
rahul = Student("Rahul", [95, 88, 79, 91])
print(f"Original marks: {rahul.marks}")

# Anyone can directly modify marks – this might be dangerous!
rahul.marks = [100, 100, 100, 100]   # directly changed
print(f"New marks: {rahul.marks}")
```


This is too permissive. What if someone accidentally sets marks to a string or an empty list? The object’s integrity is lost.

---

## 3. Protected Attributes – A Convention  

Python uses a **single underscore** before an attribute name (e.g., `_marks`) to indicate that it is **protected**.  
**This is only a convention** – the attribute is still accessible, but programmers understand it’s meant for internal use.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self._marks = marks      # protected by convention

    def total_marks(self):
        return sum(self._marks)

rahul = Student("Rahul", [95, 88, 79, 91])
print(f"Protected marks: {rahul._marks}")   # Still accessible
```


The underscore doesn’t stop anyone from changing `_marks`. It’s just a polite reminder: “This is for internal use, please don’t touch.”

---

## 4. Private Attributes – Name Mangling  

To make an attribute truly **hard to access** from outside, we use **double underscore** `__`.  
Python performs **name mangling**: it renames the attribute internally to `_ClassName__attribute`. This makes accidental access much less likely.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks      # private attribute

    def total_marks(self):
        return sum(self.__marks)

rahul = Student("Rahul", [95, 88, 79, 91])

# Attempt direct access – this will cause an AttributeError!
# print(rahul.__marks)   # Uncomment to see the error

# But we can still access it through the mangled name (not recommended!)
print(rahul._Student__marks) 
```


The double underscore makes it harder to accidentally modify the data, but determined programmers can still get to it.  
> 🧠 **Remember**: The goal of encapsulation is not to make data completely inaccessible, but to **signal intent** and prevent accidental misuse.

---

## 5. Controlling Access with Getter and Setter Methods  

The real power of encapsulation comes from providing controlled access through **methods**.  
We keep attributes private and create public methods to read (getters) and modify (setters) them. Inside these methods, we can add validation.


```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks      # private

    # Getter method – allows reading marks safely
    def get_marks(self):
        # Return a copy so the original list cannot be modified
        return self.__marks.copy()

    # Setter method – allows updating marks with validation
    def set_marks(self, new_marks):
        # Check that new_marks is a list of numbers
        if isinstance(new_marks, list) and all(isinstance(m, (int, float)) for m in new_marks):
            self.__marks = new_marks
            print("Marks updated successfully.")
        else:
            print("Error: Please provide a list of numbers.")

    def total_marks(self):
        return sum(self.__marks)

# Create a student
rahul = Student("Rahul", [95, 88, 79, 91])

# Read marks via getter
print("Current marks:", rahul.get_marks())

# Try to set invalid marks
rahul.set_marks("not a list")        # Invalid
rahul.set_marks([100, 100, 100])     # Valid (note: only 3 subjects now)
print("Updated marks:", rahul.get_marks())
```


Notice:  
- The getter returns a **copy** of the marks list, so the caller cannot accidentally change the original.  
- The setter checks the input type before updating.

---

## 6. Adding More Sophisticated Validation  

Let’s improve our setter to ensure the number of subjects matches the expected count (e.g., 6 subjects for SSLC) and that marks are between 0 and 100.


```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks      # assume marks is a list

    def get_marks(self):
        return self.__marks.copy()

    def set_marks(self, new_marks):
        # Check 1: must be a list
        if not isinstance(new_marks, list):
            print("Error: Marks must be a list.")
            return

        # Check 2: must have exactly 6 subjects (typical SSLC)
        if len(new_marks) != 6:
            print(f"Error: Expected 6 subjects, got {len(new_marks)}.")
            return

        # Check 3: all elements must be numbers between 0 and 100
        for m in new_marks:
            if not isinstance(m, (int, float)) or m < 0 or m > 100:
                print("Error: Each mark must be a number between 0 and 100.")
                return

        # All checks passed – update marks
        self.__marks = new_marks
        print("Marks updated successfully.")

    def total_marks(self):
        return sum(self.__marks)

# Test with valid and invalid data
subjects = ["Malayalam", "English", "Hindi", "Maths", "Science", "Social"]
rahul = Student("Rahul", [92, 91, 88, 95, 88, 79])
print("Original total:", rahul.total_marks())

# Try to set marks with wrong length
rahul.set_marks([100, 100, 100])   # too few

# Try marks out of range
rahul.set_marks([102, 91, 88, 95, 88, 79])   # 102 is invalid

# Correct update
rahul.set_marks([98, 91, 88, 97, 90, 85])
print("New total:", rahul.total_marks())
```

## Why Encapsulation Improves Reliability  

- **Data integrity**: Invalid data cannot sneak into your objects.  
- **Flexibility**: You can change the internal representation without breaking external code. For example, you could store marks as a dictionary internally but still provide a list via the getter.  
- **Debugging**: You can add logging or checks inside setters to track changes.  

## Summary  

- **Encapsulation** is the practice of hiding internal data and providing controlled access.  
- **Public** attributes (no underscore) are accessible everywhere.  
- **Protected** attributes (`_name`) are a convention for internal use.  
- **Private** attributes (`__name`) use name mangling to discourage direct access.  
- **Getters and setters** (or `@property`) allow you to add validation and logic when reading or writing data.  
- Encapsulation makes your code **more secure, reliable, and maintainable**.
