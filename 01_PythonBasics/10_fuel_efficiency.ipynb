# Fuel Efficiency Tracker

Hi there! In this notebook, we'll build a simple **Fuel Efficiency Tracker** to keep tabs on your vehicle's mileage and fuel costs.  
You'll practice using **lists of dictionaries**, **loops**, **functions**, and a bit of **missing data handling**.

Imagine you have a scooter or a car that you use to go to school, tuition, or trips. Every time you fill fuel, you note down:
- Liters filled
- Distance on the odometer (or distance traveled since last fill)
- Cost of the fill

By the end, you'll be able to compute your average **kilometers per liter (km/l)** and **cost per kilometer** – very handy for budgeting!



## Step 1: Store fill‑up records

We'll store each fill‑up as a **dictionary** with keys: `'liters'`, `'distance'` (km traveled since last fill), and `'cost'` (rupees).  
All records will be kept in a **list**.  

Let's create some sample data (you can replace it with your own numbers later).

```python
# List of fill-up records
fuel_log = [
    {"liters": 5.2, "distance": 120, "cost": 520},   # First fill
    {"liters": 4.8, "distance": 115, "cost": 480},   # Second fill
    {"liters": 6.0, "distance": 140, "cost": 600},   # Third fill
    {"liters": 5.5, "distance": 130, "cost": 550}    # Fourth fill
]

# Print the log to see what we have
print("Fuel Fill Log:")
for i, record in enumerate(fuel_log, start=1):
    print(f"  Fill {i}: {record}")
```


> 💡 **Tip:** `enumerate()` gives you both the index and the item – useful when printing numbered lists.

---

## Step 2: Calculate totals

Now let's compute the **total liters**, **total distance**, and **total cost** from all fill‑ups.  
We'll use a `for` loop to go through each record and add up the values.

```python
# Initialize totals
total_liters = 0
total_distance = 0
total_cost = 0

# Loop through each record
for record in fuel_log:
    total_liters += record["liters"]
    total_distance += record["distance"]
    total_cost += record["cost"]

print("--- Totals ---")
print(f"Total liters filled : {total_liters:.2f} L")
print(f"Total distance driven: {total_distance} km")
print(f"Total cost           : ₹{total_cost}")
```

---

## Step 3: Compute average mileage and cost per km

With the totals, we can find:
- **Average mileage** = total distance / total liters (km/l)
- **Cost per kilometer** = total cost / total distance (₹/km)


```python
# Calculate averages
avg_mileage = total_distance / total_liters
cost_per_km = total_cost / total_distance

print(f"Average mileage      : {avg_mileage:.2f} km/l")
print(f"Cost per kilometer   : ₹{cost_per_km:.2f} per km")
```

---

## Step 4: Wrap logic into functions

Let's create functions that can work with **any** fuel log list.  
We'll write:
- `calculate_totals(log)` → returns total liters, distance, cost
- `display_summary(log)` → prints the summary using the totals

```python
def calculate_totals(log):
    """Return (total_liters, total_distance, total_cost) for a given fuel log."""
    total_l = 0
    total_d = 0
    total_c = 0
    for entry in log:
        total_l += entry["liters"]
        total_d += entry["distance"]
        total_c += entry["cost"]
    return total_l, total_d, total_c
def display_summary(log):
    """Print a nicely formatted summary of fuel efficiency."""
    liters, dist, cost = calculate_totals(log)
    if dist == 0:   # avoid division by zero
        print("No distance recorded – cannot compute averages.")
        return
    
    mileage = dist / liters
    cost_per_km = cost / dist
    
    print("--- Fuel Efficiency Summary ---")
    print(f"Total fills          : {len(log)}")
    print(f"Total liters         : {liters:.2f} L")
    print(f"Total distance       : {dist} km")
    print(f"Total cost           : ₹{cost}")
    print(f"Average mileage      : {mileage:.2f} km/l")
    print(f"Cost per kilometer   : ₹{cost_per_km:.2f}")
    print("-------------------------------")
```

```python
# Test the function with our fuel_log
display_summary(fuel_log)
```

## Step 5: Variation – handling missing data

Sometimes you might forget to record the distance for a fill‑up.  
We can represent missing data with `None` and decide how to handle it.  
For example, if distance is `None`, we should **skip** that record when computing total distance, but still count the liters and cost? Or maybe we estimate using average? Let's keep it simple: we'll skip any record that has `None` for distance.

First, create a log with a missing distance:

```python
# Add a record with missing distance (None)
fuel_log_with_missing = [
    {"liters": 5.2, "distance": 120, "cost": 520},
    {"liters": 4.8, "distance": None, "cost": 480},   # oops, forgot distance!
    {"liters": 6.0, "distance": 140, "cost": 600},
    {"liters": 5.5, "distance": 130, "cost": 550}
]

print("Log with missing data:")
for i, rec in enumerate(fuel_log_with_missing, 1):
    print(f"  Fill {i}: {rec}")
```

Now write a new function that safely calculates totals, ignoring entries where `distance` is `None`.


```python
def calculate_totals_safe(log):
    """Return totals, skipping records with missing distance."""
    total_l = 0
    total_d = 0
    total_c = 0
    skipped = 0
    for entry in log:
        # If distance is None, skip adding to distance, but still add liters and cost? 
        # Actually, if we don't know how far we drove on that fuel, we can't use it in mileage calculation.
        # So we'll skip the whole entry for distance, but we might still want to count cost? 
        # For simplicity, we'll skip the entry entirely for all totals – otherwise mileage would be wrong.
        # But cost per km would also be affected. A better approach: only include records that have both distance and liters.
        # We'll check if distance is not None and also ensure liters is not None (though liters should always be known).
        if entry["distance"] is not None and entry["liters"] is not None:
            total_l += entry["liters"]
            total_d += entry["distance"]
            total_c += entry["cost"]
        else:
            skipped += 1
    print(f"(Skipped {skipped} record(s) due to missing data)")
    return total_l, total_d, total_c

# Test with the log that has missing data
liters, dist, cost = calculate_totals_safe(fuel_log_with_missing)
print(f"Totals after skipping: liters={liters:.2f}, distance={dist}, cost={cost}")

# Now compute average only if distance > 0
if dist > 0:
    mileage = dist / liters
    cost_per_km = cost / dist
    print(f"Average mileage: {mileage:.2f} km/l")
    print(f"Cost per km    : ₹{cost_per_km:.2f}")
else:
    print("Not enough valid data to compute averages.")
```


> **Note:** We used `is not None` to check for missing values. `None` is a special value in Python that means "nothing" or "no value".
