# String Operations

### 1. **String Concatenation**
Combining two or more strings.

```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)  # Output: John Doe
```

---

### 2. **String Interpolation (f-strings)**
Embedding expressions inside string literals.

```python
name = "Alice"
age = 25
greeting = f"My name is {name} and I am {age} years old."
print(greeting)  # Output: My name is Alice and I am 25 years old.
```

---

### 3. **String Formatting**
Using the `.format()` method.

```python
name = "Bob"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)  # Output: My name is Bob and I am 30 years old.
```

---

### 4. **String Slicing**
Extracting a portion of a string.

```python
text = "Hello, World!"
print(text[0:5])  # Output: Hello
print(text[7:])   # Output: World!
print(text[-6:-1])  # Output: World
```

---

### 5. **String Length**
Getting the length of a string.

```python
text = "Python"
print(len(text))  # Output: 6
```

---

### 6. **String Case Conversion**
Changing the case of a string.

```python
text = "Hello, World!"
print(text.upper())  # Output: HELLO, WORLD!
print(text.lower())  # Output: hello, world!
print(text.title())  # Output: Hello, World!
```

---

### 7. **String Stripping**
Removing leading and trailing whitespace.

```python
text = "   Python   "
print(text.strip())  # Output: Python
print(text.lstrip())  # Output: Python   (left strip)
print(text.rstrip())  # Output:    Python (right strip)
```

---

### 8. **String Splitting**
Splitting a string into a list.

```python
text = "apple,banana,cherry"
fruits = text.split(",")
print(fruits)  # Output: ['apple', 'banana', 'cherry']
```

---

### 9. **String Joining**
Combining a list of strings into a single string.

```python
fruits = ["apple", "banana", "cherry"]
text = ", ".join(fruits)
print(text)  # Output: apple, banana, cherry
```

---

### 10. **String Replacement**
Replacing substrings.


# Integers

In Python, integers (`int`) are a fundamental data type used to represent whole numbers. While integers themselves don't have as many operations as strings, there are still a variety of operations and methods you can perform with them. Below is a comprehensive list of **integer operations**:

---

### 1. **Basic Arithmetic Operations**
   - **Addition (`+`)**: Adds two integers.
     ```python
     a = 5
     b = 3
     print(a + b)  # Output: 8
     ```
   - **Subtraction (`-`)**: Subtracts one integer from another.
     ```python
     print(a - b)  # Output: 2
     ```
   - **Multiplication (`*`)**: Multiplies two integers.
     ```python
     print(a * b)  # Output: 15
     ```
   - **Division (`/`)**: Divides one integer by another (returns a float).
     ```python
     print(a / b)  # Output: 1.6666666666666667
     ```
   - **Floor Division (`//`)**: Divides and returns the integer part of the result.
     ```python
     print(a // b)  # Output: 1
     ```
   - **Modulus (`%`)**: Returns the remainder of division.
     ```python
     print(a % b)  # Output: 2
     ```
   - **Exponentiation (`**`)**: Raises one integer to the power of another.
     ```python
     print(a ** b)  # Output: 125
     ```

---

### 2. **Bitwise Operations**
   - **AND (`&`)**: Performs a bitwise AND operation.
     ```python
     print(a & b)  # Output: 1 (5 & 3 = 0b101 & 0b011 = 0b001)
     ```
   - **OR (`|`)**: Performs a bitwise OR operation.
     ```python
     print(a | b)  # Output: 7 (5 | 3 = 0b101 | 0b011 = 0b111)
     ```
   - **XOR (`^`)**: Performs a bitwise XOR operation.
     ```python
     print(a ^ b)  # Output: 6 (5 ^ 3 = 0b101 ^ 0b011 = 0b110)
     ```
   - **NOT (`~`)**: Performs a bitwise NOT operation (inverts bits).
     ```python
     print(~a)  # Output: -6 (~5 = -6)
     ```
   - **Left Shift (`<<`)**: Shifts bits to the left.
     ```python
     print(a << 1)  # Output: 10 (5 << 1 = 0b101 << 1 = 0b1010)
     ```
   - **Right Shift (`>>`)**: Shifts bits to the right.
     ```python
     print(a >> 1)  # Output: 2 (5 >> 1 = 0b101 >> 1 = 0b10)
     ```

---

### 3. **Comparison Operations**
   - **Equal (`==`)**: Checks if two integers are equal.
     ```python
     print(a == b)  # Output: False
     ```
   - **Not Equal (`!=`)**: Checks if two integers are not equal.
     ```python
     print(a != b)  # Output: True
     ```
   - **Greater Than (`>`)**: Checks if one integer is greater than another.
     ```python
     print(a > b)  # Output: True
     ```
   - **Less Than (`<`)**: Checks if one integer is less than another.
     ```python
     print(a < b)  # Output: False
     ```
   - **Greater Than or Equal (`>=`)**: Checks if one integer is greater than or equal to another.
     ```python
     print(a >= b)  # Output: True
     ```
   - **Less Than or Equal (`<=`)**: Checks if one integer is less than or equal to another.
     ```python
     print(a <= b)  # Output: False
     ```

---

### 4. **Assignment Operations**
   - **Assign (`=`)**: Assigns a value to a variable.
     ```python
     x = 10
     print(x)  # Output: 10
     ```
   - **Add and Assign (`+=`)**: Adds and assigns the result.
     ```python
     x += 5
     print(x)  # Output: 15
     ```
   - **Subtract and Assign (`-=`)**: Subtracts and assigns the result.
     ```python
     x -= 3
     print(x)  # Output: 12
     ```
   - **Multiply and Assign (`*=`)**: Multiplies and assigns the result.
     ```python
     x *= 2
     print(x)  # Output: 24
     ```
   - **Divide and Assign (`/=`)**: Divides and assigns the result.
     ```python
     x /= 4
     print(x)  # Output: 6.0
     ```
   - **Floor Divide and Assign (`//=`)**: Floor divides and assigns the result.
     ```python
     x //= 2
     print(x)  # Output: 3.0
     ```
   - **Modulus and Assign (`%=`)**: Takes modulus and assigns the result.
     ```python
     x %= 2
     print(x)  # Output: 1.0
     ```
   - **Exponentiate and Assign (`**=`)**: Raises to the power and assigns the result.
     ```python
     x **= 3
     print(x)  # Output: 1.0
     ```

---

### 5. **Type Conversion**
   - Convert to `float`:
     ```python
     x = 10
     y = float(x)
     print(y)  # Output: 10.0
     ```
   - Convert to `str`:
     ```python
     z = str(x)
     print(z)  # Output: "10"
     ```
   - Convert to `bool`:
     ```python
     print(bool(x))  # Output: True (non-zero integers are True)
     print(bool(0))  # Output: False (zero is False)
     ```

---

### 6. **Mathematical Functions**
   - **Absolute Value (`abs()`)**: Returns the absolute value of an integer.
     ```python
     print(abs(-5))  # Output: 5
     ```
   - **Power (`pow()`)**: Raises an integer to a power.
     ```python
     print(pow(2, 3))  # Output: 8
     ```
   - **Round (`round()`)**: Rounds a floating-point number to the nearest integer.
     ```python
     print(round(3.6))  # Output: 4
     ```
   - **Maximum (`max()`)**: Returns the largest of two or more integers.
     ```python
     print(max(5, 10, 3))  # Output: 10
     ```
   - **Minimum (`min()`)**: Returns the smallest of two or more integers.
     ```python
     print(min(5, 10, 3))  # Output: 3
     ```

---

### 7. **Integer Methods**
   - **Bit Length (`bit_length()`)**: Returns the number of bits required to represent the integer in binary.
     ```python
     x = 10
     print(x.bit_length())  # Output: 4 (10 in binary is 1010)
     ```
   - **Conversion to Binary (`bin()`)**: Converts an integer to a binary string.
     ```python
     print(bin(x))  # Output: '0b1010'
     ```
   - **Conversion to Hexadecimal (`hex()`)**: Converts an integer to a hexadecimal string.
     ```python
     print(hex(x))  # Output: '0xa'
     ```
   - **Conversion to Octal (`oct()`)**: Converts an integer to an octal string.
     ```python
     print(oct(x))  # Output: '0o12'
     ```

---

### 8. **Special Integer Operations**
   - **Factorial (`math.factorial()`)**: Computes the factorial of an integer.
     ```python
     import math
     print(math.factorial(5))  # Output: 120
     ```
   - **GCD (`math.gcd()`)**: Computes the greatest common divisor of two integers.
     ```python
     print(math.gcd(12, 18))  # Output: 6
     ```
